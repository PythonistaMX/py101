# Entornos virtuales.

Cuando se tienen varios proyectos de desarrollo en Python, es conveniente separar las bibliotecas de cada proyecto de las bibliotecas principales del sistema. Esto evita conflictos entre versiones y facilita reproducir entornos de trabajo independientes.

## Directorios de búsqueda de bibliotecas de módulos.

Como se discutió previamente, Python define rutas en las cuales busca bibliotecas y módulos. Esas rutas pueden consultarse, entre otras formas, mediante ```sys.path```.

````python
from sys import path  
path
````

Los entornos virtuales permiten trabajar con una instalación aislada de paquetes para un proyecto concreto, sin alterar la instalación principal de Python del sistema.

### El módulo ```venv```.

A partir de Python 3.3, la biblioteca estándar incluye el módulo ```venv```, que permite crear entornos virtuales sin instalar paquetes adicionales.

Esta es la opción básica recomendada cuando se trabaja con el intérprete estándar de Python.

* La siguiente celda creará un entorno virtual llamado ```prueba```. El directorio se crea automáticamente si no existe.

````python
!python3 -m venv prueba
````

### Habilitando el entorno virtual.

Para habilitar un entorno virtual, es necesario modificar variables de entorno desde una terminal. Esos cambios solo tienen efecto en la terminal actual.

#### Para Linux y macOS:
```bash
source <directorio>/bin/activate
```

#### Para Windows en PowerShell:

```powershell
.\<directorio>\Scripts\Activate.ps1
```

#### Para Windows en cmd:

```bat
<directorio>\Scripts\activate.bat
```

Una vez activado el entorno, las instalaciones realizadas con ```pip``` quedarán dentro de ese entorno y no en la biblioteca principal del sistema.

Esto se relaciona directamente con el control de dependencias: cada proyecto puede mantener sus propias versiones de paquetes sin interferir con otros proyectos.

**Ejemplo:**

Para activar el entorno virtual localizado en [```prueba```](prueba) es necesario ejecutar el comando correspondiente desde una terminal ubicada en el directorio del proyecto.

Para conocer el directorio en el que se encuentra la notebook se utiliza el siguiente comando mágico de Jupyter:

````python
%pwd
````

* Para Linux y MacOS X

``` bash 
$ source prueba/bin/activate
```

* Para Windows en PowerShell:

```powershell
$ .\prueba\Scripts\Activate.ps1
```

* Para Windows en cmd:

```bat
prueba\Scripts\activate.bat
```

**Advertencia:**

Activar un entorno virtual modifica el estado de la terminal, no el de una celda aislada de Jupyter. Por ello, intentar ejecutar ```source``` o comandos equivalentes dentro de una celda no produce el mismo efecto que en una terminal interactiva.

En notebooks, lo normal es seleccionar el intérprete o kernel correspondiente al entorno virtual en lugar de intentar activarlo dentro de una celda.

### Deshabilitando el entorno virtual.

Para salir del entorno virtual se utiliza el comando ```deactivate``` en cualquier plataforma.

**Nota:** El comando ```deactivate``` no es un ejecutable independiente, sino una función o script auxiliar disponible después de activar el entorno.

### Alternativas modernas.

Además de ```venv```, herramientas como ```uv``` también pueden crear y administrar entornos virtuales, a menudo integrándolos con la instalación y sincronización de dependencias.

En un flujo moderno, el entorno virtual y el control de dependencias suelen gestionarse juntos.
