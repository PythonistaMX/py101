# Entornos virtuales.

Cuando se tienen varios proyectos de desarrollo en *Python*, es conveniente separar las bibliotecas de cada proyecto de las bibliotecas principales del sistema. Esto evita conflictos entre versiones y facilita reproducir entornos de trabajo independientes.

## Directorios de búsqueda de bibliotecas de módulos.

Como se discutió previamente, *Python* define rutas en las cuales busca bibliotecas y módulos. Esas rutas pueden consultarse, entre otras formas, mediante ```sys.path```.

````python
from sys import path  
path
````

Los entornos virtuales permiten trabajar con una instalación aislada de paquetes para un proyecto concreto, sin alterar la instalación principal de *Python* del sistema.

### El módulo ```venv```.

A partir de *Python* 3.3, la biblioteca estándar incluye el módulo ```venv```, que permite crear entornos virtuales sin instalar paquetes adicionales.

Esta es la opción básica recomendada cuando se trabaja con el intérprete estándar de *Python*.

* La siguiente celda creará un entorno virtual llamado ```prueba```. El directorio se crea automáticamente si no existe.

````python
!python3 -m venv prueba
````

### Habilitando el entorno virtual.

Para habilitar un entorno virtual, es necesario modificar variables de entorno desde una terminal. Esos cambios solo tienen efecto en la terminal actual.

#### Para *GNU/Linux* y *macOS*:
```bash
source <directorio>/bin/activate
```

#### Para *Windows* en *PowerShell*:

```powershell
.\\<directorio>\\Scripts\\Activate.ps1
```

#### Para *Windows* en cmd:

```bat
<directorio>\\Scripts\\activate.bat
```

Una vez activado el entorno, las instalaciones realizadas con ```pip``` quedarán dentro de ese entorno y no en la biblioteca principal del sistema.

Esto se relaciona directamente con el control de dependencias: cada proyecto puede mantener sus propias versiones de paquetes sin interferir con otros proyectos.

**Ejemplo:**

Para activar el entorno virtual localizado en [```prueba```](prueba) es necesario ejecutar el comando correspondiente desde una terminal ubicada en el directorio del proyecto.

Para conocer el directorio en el que se encuentra la *notebook* se utiliza el siguiente comando mágico de *Jupyter*:

````python
%pwd
````

* Para *GNU/Linux* y *macOS*.

``` bash 
$ source prueba/bin/activate
```

* Para *Windows* en *PowerShell*:

```powershell
$ .\\prueba\\Scripts\\Activate.ps1
```

* Para *Windows* en cmd:

```bat
prueba\\Scripts\\activate.bat
```

**Advertencia:**

Activar un entorno virtual modifica el estado de la terminal, no el de una celda aislada de *Jupyter*. Por ello, intentar ejecutar ```source``` o comandos equivalentes dentro de una celda no produce el mismo efecto que en una terminal interactiva.

En *notebooks*, lo normal es seleccionar el intérprete o *kernel* correspondiente al entorno virtual en lugar de intentar activarlo dentro de una celda.

### Deshabilitando el entorno virtual.

Para salir del entorno virtual se utiliza el comando ```deactivate``` en cualquier plataforma.

**Nota:** El comando ```deactivate``` no es un ejecutable independiente, sino una función o *script* auxiliar disponible después de activar el entorno.


## Otras herramientas para gestión de entornos y paquetes.

Además del módulo estándar `venv`, existe un ecosistema amplio de herramientas especializadas para la gestión de entornos virtuales y paquetes en *Python*. Cada una responde a necesidades distintas de los proyectos y flujos de trabajo.


### Conda, Anaconda y Miniconda.

*Conda* es un sistema de gestión de entornos y paquetes de código abierto, muy popular en el ámbito de la ciencia de datos, el aprendizaje automático y el cómputo científico.

A diferencia de `venv` y `pip`, *conda* no está limitado a *Python*: puede gestionar paquetes de cualquier lenguaje (R, C, C++, etc.) y sus dependencias binarias nativas.

**Distribuciones principales:**

| Distribución | Descripción |
|---|---|
| **Anaconda** | Distribución completa que incluye *Python*, *conda* y más de 250 paquetes científicos preinstalados (*NumPy*, *Pandas*, *Matplotlib*, *Jupyter*, etc.). Ocupa aproximadamente 3 GB. Recomendada para quienes se inician en ciencia de datos. |
| **Miniconda** | Versión mínima que solo instala *Python* y *conda*. Los paquetes adicionales se instalan según se necesiten. Recomendada para usuarios con más experiencia. |
| **Miniforge** | Alternativa comunitaria a *Miniconda* que configura por defecto el canal `conda-forge`, el repositorio de paquetes más completo de la comunidad. |

La instalación se realiza descargando el instalador desde:
- Anaconda: [https://www.anaconda.com](https://www.anaconda.com)
- Miniconda: [https://docs.conda.io/en/latest/miniconda.html](https://docs.conda.io/en/latest/miniconda.html)
- Miniforge: [https://github.com/conda-forge/miniforge](https://github.com/conda-forge/miniforge)


**Comandos básicos de *conda*.**

```bash
# Crear un entorno virtual con una versión específica de Python
conda create --name mi_entorno python=3.12

# Activar el entorno
conda activate mi_entorno

# Desactivar el entorno
conda deactivate

# Listar los entornos disponibles
conda env list

# Instalar un paquete en el entorno activo
conda install numpy

# Instalar desde el canal conda-forge
conda install -c conda-forge pandas

# Exportar el entorno a un archivo YAML para reproducibilidad
conda env export > entorno.yml

# Recrear un entorno a partir de un archivo YAML
conda env create -f entorno.yml

# Eliminar un entorno
conda remove --name mi_entorno --all
```

**Nota:** *conda* permite mezclar paquetes de sus propios repositorios con paquetes instalados vía `pip`. Sin embargo, se recomienda preferir `conda` siempre que el paquete esté disponible en sus canales, para evitar conflictos de dependencias.


### pyenv.

*pyenv* es una herramienta que permite instalar y cambiar entre múltiples versiones de *Python* en el mismo sistema, sin interferir con la instalación del sistema operativo.

Es especialmente útil cuando se trabaja con proyectos que requieren versiones distintas de *Python*.

```bash
# Listar las versiones de Python disponibles para instalar
pyenv install --list

# Instalar una versión específica de Python
pyenv install 3.12.0

# Establecer la versión global del sistema
pyenv global 3.12.0

# Establecer la versión local para un directorio de proyecto
# (crea un archivo .python-version en el directorio)
pyenv local 3.11.5

# Listar las versiones instaladas
pyenv versions

# Ver la versión activa en el contexto actual
pyenv version
```

*pyenv* se combina habitualmente con `venv` para crear entornos aislados por proyecto. También puede usarse con el plugin `pyenv-virtualenv` para unificar ambas funciones.

**Instalación:** `pyenv` no está disponible en *Windows* de forma nativa; en ese sistema se utiliza *pyenv-win* ([https://github.com/pyenv-win/pyenv-win](https://github.com/pyenv-win/pyenv-win)).


### pipenv.

*pipenv* combina la gestión de paquetes (`pip`) con la gestión de entornos virtuales (`venv`) en una sola herramienta. Elimina la necesidad de gestionar manualmente archivos `requirements.txt` y el entorno virtual por separado.

Utiliza dos archivos para describir las dependencias del proyecto:

- **`Pipfile`**: declara las dependencias directas del proyecto (similar a `requirements.txt` pero en formato *TOML*). Distingue entre dependencias de producción y de desarrollo.
- **`Pipfile.lock`**: registra las versiones exactas de todos los paquetes instalados (incluyendo dependencias transitivas), garantizando instalaciones reproducibles.

```bash
# Instalar pipenv
pip install pipenv

# Crear un entorno e instalar una dependencia de producción
pipenv install requests

# Instalar una dependencia solo para desarrollo
pipenv install pytest --dev

# Activar el entorno virtual del proyecto
pipenv shell

# Ejecutar un script dentro del entorno sin activarlo
pipenv run python script.py

# Instalar todas las dependencias declaradas en el Pipfile
pipenv install

# Instalar también las dependencias de desarrollo
pipenv install --dev

# Ver el grafo de dependencias
pipenv graph
```


### Poetry.

*Poetry* es una herramienta moderna de gestión de dependencias y empaquetado para *Python*. Usa el archivo `pyproject.toml` (estándar *PEP 518* y *PEP 621*) para centralizar toda la configuración del proyecto: dependencias, metadatos, scripts y configuración de herramientas.

Ofrece resolución determinista de dependencias mediante un archivo `poetry.lock` y facilita la publicación de paquetes en *PyPI*.

```bash
# Instalar Poetry
pip install poetry

# Crear un nuevo proyecto con la estructura recomendada
poetry new mi_proyecto

# Inicializar Poetry en un proyecto existente
poetry init

# Agregar una dependencia de producción
poetry add requests

# Agregar una dependencia de desarrollo
poetry add pytest --group dev

# Instalar todas las dependencias del proyecto
poetry install

# Activar el entorno virtual gestionado por Poetry
poetry shell

# Ejecutar un comando dentro del entorno sin activarlo
poetry run python script.py

# Actualizar las dependencias a las últimas versiones permitidas
poetry update

# Publicar el paquete en PyPI
poetry publish --build
```

*Poetry* es especialmente recomendado para el desarrollo de bibliotecas y paquetes que se distribuirán en *PyPI*.


### uv.

*uv* es una herramienta de gestión de paquetes y entornos virtuales extremadamente rápida, escrita en *Rust*. Fue desarrollada por *Astral* (los creadores de *ruff*) y está diseñada como un reemplazo compatible con `pip`, `pip-tools`, `pipx`, `virtualenv` y, parcialmente, con *Poetry*.

Su principal ventaja es la velocidad: en la mayoría de los casos resuelve e instala dependencias entre 10 y 100 veces más rápido que `pip`.

```bash
# Instalar uv
pip install uv

# Crear un entorno virtual
uv venv

# Crear un entorno virtual con una versión específica de Python
uv venv --python 3.12

# Activar el entorno (igual que con venv)
source .venv/bin/activate        # Linux / macOS
.venv\Scripts\activate           # Windows

# Instalar paquetes (equivalente a pip install)
uv pip install requests numpy

# Instalar desde un archivo requirements.txt
uv pip install -r requirements.txt

# Sincronizar el entorno con un requirements.txt (elimina paquetes no listados)
uv pip sync requirements.txt

# Generar un requirements.txt con versiones fijadas
uv pip compile pyproject.toml -o requirements.txt
```

*uv* también incluye un gestor de proyectos completo (similar a *Poetry*) mediante los subcomandos `uv init`, `uv add`, `uv run` y `uv sync`, usando `pyproject.toml` como archivo de configuración.


## Comparativa de herramientas.

La siguiente tabla resume las capacidades de las herramientas más utilizadas para la gestión de entornos y paquetes en *Python*.

| Herramienta | Gestión de versiones de Python | Entornos virtuales | Gestión de paquetes | Empaquetado y publicación | Caso de uso principal |
|---|:---:|:---:|:---:|:---:|---|
| `venv` + `pip` | No | Sí | Sí | No | Uso general, sin dependencias extra |
| `conda` / Anaconda | Sí | Sí | Sí | Parcial | Ciencia de datos, paquetes con binarios nativos |
| `pyenv` | Sí | No* | No | No | Cambiar entre versiones de Python |
| `pipenv` | No | Sí | Sí | No | Aplicaciones, separación dev/prod |
| `poetry` | No | Sí | Sí | Sí | Bibliotecas y paquetes publicables |
| `uv` | Sí** | Sí | Sí | Sí** | Máxima velocidad, reemplazo moderno de pip |

\* *pyenv* puede gestionar entornos con el plugin `pyenv-virtualenv`.  
\*\* *uv* puede instalar versiones de *Python* y gestionar proyectos completos a partir de la versión 0.4.

**Recomendaciones generales:**

- **Ciencia de datos y aprendizaje automático**: usar *Miniconda* o *Miniforge* con *conda*.
- **Desarrollo de aplicaciones**: *pipenv* o *poetry* aportan mayor control sobre las dependencias.
- **Desarrollo de bibliotecas distribuibles en PyPI**: *poetry* o *uv* (modo proyecto).
- **Necesidad de múltiples versiones de Python**: combinar *pyenv* con `venv` o *pyenv-virtualenv*.
- **Velocidad e integración en CI/CD**: *uv* es la opción más moderna y rápida.
