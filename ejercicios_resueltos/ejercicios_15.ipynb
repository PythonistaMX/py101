#### Ejercicio 1. Función que reciba una lista de números y devuelva el número más grande y el más pequeño.

Defina una función `encontrar_extremos` que reciba una lista de números y devuelva una tupla con el número más grande y el número más pequeño de la lista sin utilizar las funciones `max` y `min`. Si la lista está vacía, la función debe indicarlo.

````python
def encontrar_extremos(lista):
    if not lista:
        return 'La lista no puede estar vacía.'

    maximo = lista[0] # Inicializamos el máximo con el primer elemento de la lista.
    minimo = lista[0] # Inicializamos el mínimo con el primer elemento de la lista.
    for i in range(1, len(lista)):
        if lista[i] > maximo: # Si encontramos un número mayor que el máximo actual, lo actualizamos.
            maximo = lista[i]
        if lista[i] < minimo: # Si encontramos un número menor que el mínimo actual, lo actualizamos.
            minimo = lista[i]
    return (maximo, minimo) # Devolvemos una tupla con el máximo y el mínimo encontrados.
````

````python
encontrar_extremos([32, 11, 24, 1, -23, 4000, 32])
````

#### Ejercicio 2. Función que reciba una cantidad arbitraria de números y devuelva el número más grande y el más pequeño.

Defina una función `encontrar_extremos_varios` que reciba una cantidad arbitraria de números (utilizando `*args`) y devuelva una tupla con el número más grande y el número más pequeño de los argumentos recibidos sin utilizar las funciones `max` y `min`. Si no se recibe ningún argumento, la función debe indicarlo.

````python
def encontrar_extremos_varios(*args):
    if not args:
        return 'Se requiere al menos un número.'

    maximo = args[0] # Inicializamos el máximo con el primer argumento recibido.
    minimo = args[0] # Inicializamos el mínimo con el primer argumento recibido.
    for i in range(1, len(args)):
        if args[i] > maximo: # Si encontramos un número mayor que el máximo actual, lo actualizamos.
            maximo = args[i]
        if args[i] < minimo: # Si encontramos un número menor que el mínimo actual, lo actualizamos.
            minimo = args[i]
    return (maximo, minimo) # Devolvemos una tupla con el máximo y el mínimo encontrados.
````

````python
encontrar_extremos_varios(32, 11, 24, 1, -23, 4000, 32)
````

#### Ejercicio 3. Función que cree una figura triangular, definiendo la altura y el carácter a utilizar.
Defina una función `triangulo` que reciba un número entero `altura` y un carácter `caracter` y que imprima un triángulo de la altura dada utilizando el carácter especificado. Por ejemplo, `triangulo(4, '*')` debería imprimir:

```
   *
  ***
 *****
*******
 ```

````python
def triangulo(altura=5, caracter='*'):
    for i in range(1, altura + 1):
        espacios = ' ' * (altura - i)
        caracteres = caracter * (2 * i - 1)
        print(espacios + caracteres) # Imprime una línea con los espacios y el carácter formando un triángulo.
````

#### Ejercicio 4. Función que calcule la frecuencia de cada carácter en una cadena.

Defina una función `frecuencia_caracteres` que reciba una cadena de texto y devuelva un diccionario donde las claves sean los caracteres únicos de la cadena y los valores sean la cantidad de veces que cada carácter aparece en la cadena. Por ejemplo, `frecuencia_caracteres("hello")` debería devolver `{'h': 1, 'e': 1, 'l': 2, 'o': 1}`.

````python
def frecuencia_caracteres(cadena):
    frecuencia = {} # Creamos un diccionario para contar la frecuencia de cada carácter.
    for char in cadena:
        if char in frecuencia:
            frecuencia[char] += 1 # Si el carácter ya está en el diccionario, incrementamos su contador.
        else:
            frecuencia[char] = 1 # Si el carácter no está en el diccionario, lo añadimos con un contador de 1.
    return frecuencia # Devolvemos el diccionario con la frecuencia de cada carácter.
````

````python
frecuencia_caracteres('pythonista') # Llama a la función con la cadena 'pythonista'.
````

#### Ejercicio 5. Función que regresa una cadena ordenada alfabéticamente.

Usando la función `frecuencia_caracteres()` del ejercicio anterior, defina una función `ordenar_cadena()` que reciba una cadena de texto y devuelva una nueva cadena con los caracteres ordenados alfabéticamente de menor a mayor, conservando tantas repeticiones como aparezcan en la cadena original.

Además debe contener el parámetro opcional `reversa` que, si se establece en `True`, ordenará los caracteres de `z` a `a`. Por ejemplo, `ordenar_cadena("banana")` debería devolver `"aaabnn"`.

Puede utilizar la función `sorted()` para ordenar los caracteres únicos, pero no debe usar el método `str.join()` para construir la cadena resultante.

````python
def ordenar_cadena(cadena, reversa=False):
    frecuencias = frecuencia_caracteres(cadena) # Obtenemos la frecuencia de cada carácter en la cadena.
    cadena_unicos = ''
    for char in frecuencias:
        cadena_unicos += char # Construimos una cadena con los caracteres únicos de la cadena original.
    cadena_ordenada = sorted(cadena_unicos, reverse=reversa) # Ordenamos los caracteres únicos alfabéticamente.
    cadena_final = ''
    for char in cadena_ordenada:
        cadena_final += char * frecuencias[char] # Repetimos cada carácter según su frecuencia original.
    return cadena_final # Devolvemos la cadena ordenada.
````

````python
ordenar_cadena('pythonista', reversa=True) # Llama a la función con la cadena 'pythonista'.
````
