## Ejercicios tipo entrevista.

Los siguientes ejercicios son demostrativos, con preguntas breves y soluciones directas.

#### Ejercicio 1. Búsqueda de anagramas.

Usando una expresión, evalúe si dos palabras son anagramas, es decir, si contienen las mismas letras en diferente orden.

Evalúe las palabras:
* "amor" y "roma".

```python
# La siguiente expresión utiliza la función `sorted()` para ordenar 
# las letras de cada palabra y luego compara si son iguales.
# Si las palabras son anagramas, sus letras ordenadas serán iguales.
sorted("amor") == sorted("roma")
```

#### Ejercicio 2. Encuentre el número más grande.

Dado el siguiente conjunto de números, encuentre el número más grande usando una función de *Python*.
* 12, 45, 67, 23, 89, 34

```python
# La función `max()` devuelve el valor máximo de los argumentos proporcionados.
max(12, 45, 67, 23, 89, 34)
```

#### Ejercicio 3. Replicación de una lista.

Dado el siguiente conjunto de números, cree una nueva lista que contenga los mismos elementos que la lista original.
* [1, 2, 3, 4, 5] 

```python
# El rebanado de listas para todo el contenido de una lista crea una copia superficial.
lista_1 = [1, 2, 3, 4, 5]
lista_2 = lista_1[:]
```

Esto se comprueba de la siguiente manera:

```python
# El operador `==` compara el contenido de las listas.
print("¿Ambas listas contienen los mismos elementos?:", lista_1 == lista_2)
# El operador `is` compara si ambas variables apuntan al mismo objeto en memoria.
print("¿Ambas listas son el mismo objeto en memoria?:", lista_1 is lista_2)
# Las funciones `id()` muestran la dirección de memoria de cada lista.
print("ID de lista_1:", id(lista_1))
print("ID de lista_2:", id(lista_2))
```

#### Ejercicio 4. Números más grandes en una tupla.

Regrese los 3 números más grandes de la tupla:
* (10, 20, 5, 30, 15, 25)

```python
# La función `sorted()` ordena los números en orden descendente
# mediante el argumento `reverse=True`, y luego se seleccionan
# los primeros tres elementos.
sorted((10, 20, 5, 30, 15, 25), reverse=True)[:3]
```

#### Ejercicio 5. Obtener el inverso de una cadena.

Dada la cadena "*Python*", obtenga la cadena invertida.

```python
# Las colecciones ordenadas como listas y tuplas y
# cadenas de caracteres pueden ser rebanadas utilizando
# la sintaxis de rebanado [::-1].
"Python"[::-1]
```

#### Ejercicio 6. Evaluar si la longitud de una cadena es par o impar.

Usando sólo una expresión, evalúe si la longitud de la cadena "Monty" es par o impar, regresando "par" o "impar" en consecuencia.

```python
# La función `len()` devuelve la longitud de la cadena "Monty", que es 5.
# El operador módulo `%` calcula el resto de la división de 5 entre 2, que es 1.
# La expresión completa evalúa si el resultado es igual a 0, lo que 
# indicaría que la longitud es par.
"par" if len("Monty") % 2 == 0 else "impar"
```

#### Ejercicio 7. Evaluar si una palabra es un palíndromo.

Usando una expresión, evalúe si la palabra "radar" es un palíndromo, es decir, si se lee igual de izquierda a derecha y de derecha a izquierda.

```python
# La expresión compara la cadena con su reverso utilizando el rebanado [::-1].
"radar" == "radar"[::-1]
```

#### Ejercicio 8. Evaluar si las variables están vacías.
Las siguientes variables contienen distintos tipos de datos. Evalúe si cada una de ellas está vacía o no, regresando "vacía" o "no vacía" en consecuencia.
* `cadena_1 = "Hola"`
* `cadena_2 = ""`
* `lista_1 = []`
* `lista_2 = [1, 2, 3]`
* `valor_none = None`
* `tupla_2 = (1, 2, 3)`
* `tupla_3 = (None,)`
* `conjunto = set()`
* `diccionario = {"contenido": ""}`

```python
cadena_1 = "Hola"
cadena_2 = ""
lista_1 = []   
lista_2 = [1, 2, 3]
valor_none = None
tupla_2 = (1, 2, 3)
tupla_3 = (None,)
conjunto = set()
diccionario = {"contenido": ""}

# Una colección vacía se considera `False` en un contexto booleano.
# Las f-strings permiten incluir expresiones dentro de cadenas de texto utilizando llaves `{}`.
# El operador ternario dentro de las f-strings evalúa si cada valor es vacío o no
# y devuelve el mensaje correspondiente.
print(f"'{cadena_1}', {'no está vacía.' if bool(cadena_1) else 'está vacía.'}")
print(f"'{cadena_2}', {'no está vacía.' if bool(cadena_2) else 'está vacía.'}")
print(f"{lista_1}, {'no está vacía.' if bool(lista_1) else 'está vacía.'}")
print(f"{lista_2}, {'no está vacía.' if bool(lista_2) else 'está vacía.'}")
# `None` representa ausencia de valor y se evalúa como `False`.
print(f"{valor_none}, {'no está vacía.' if bool(valor_none) else 'está vacía.'}")
print(f"{tupla_2}, {'no está vacía.' if bool(tupla_2) else 'está vacía.'}")
# `(None,)` crea una tupla de un elemento, por eso no es vacía.
print(f"{tupla_3}, {'no está vacía.' if bool(tupla_3) else 'está vacía.'}")
print(f"{conjunto}, {'no está vacía.' if bool(conjunto) else 'está vacía.'}")
print(f"{diccionario}, {'no está vacía.' if bool(diccionario) else 'está vacía.'}")
```

#### Ejercicio 9. Transformar un mensaje en binario a texto.

#### Ejercicio 10. Revisión de codificación de caracteres.

Al intentar convertir la siguiente cadena `bytes` a una cadena de texto usando `str(cadena_bytes, encoding='utf-8')`, se produce un error de decodificación. Identifique el error y proponga una solución para convertir correctamente la cadena a texto.

* `b'Kurt Friedrich G\xf6del fue un matem\xe1tico y fil\xf3sofo.'`

```python
# El resultado de la conversión de bytes usando `str()` con la codificación 'utf-8'
# regresa un mensaje de error.

cadena_bytes = b'Kurt Friedrich G\xf6del fue un matem\xe1tico y fil\xf3sofo.'
str(cadena_bytes, encoding='utf-8')
```

```python
# Es necesario usar la codificación correcta para decodificar los bytes, en este caso 'latin-1' o 'iso-8859-1'.
str(cadena_bytes, encoding='latin-1')
```
