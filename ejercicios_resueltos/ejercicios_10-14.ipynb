#### Ejercicio 1. Algoritmo para obtener el valor máximo de una lista de números.

Sin usar la función `max()`, escriba un algoritmo para obtener el valor máximo de una lista de números.

*Nota:* Este ejercicio puede resolverse mediante un recorrido lineal, conservando en una variable el mayor valor encontrado hasta cada iteración.

````python
lista = [32, 11, 24, 1, -23, 4000, 32]
maximo = lista[0] # Inicializamos el máximo con el primer elemento de la lista.
for num in lista:
    if num > maximo: # Si encontramos un número mayor que el máximo actual, lo actualizamos.
        maximo = num
print(maximo) # Imprimimos el número máximo encontrado en la lista.
````

#### Ejercicio 2. Algoritmo para ordenar una lista de números.

Sin usar las funciones `sort()`, `max()`, `min()` o métodos predefinidos, escriba un algoritmo para ordenar una lista de números de menor a mayor.

*Nota:* Este algoritmo se llama ["ordenamiento por selección"](https://es.wikipedia.org/wiki/Ordenamiento_por_selecci%C3%B3n) y es uno de los algoritmos de ordenamiento más simples, aunque no es el más eficiente para listas grandes.

````python
lista = [32, 11, 24, 1, -23, 4000, 32]
lista_ordenada = []
maximo = lista[0] # Inicializamos el máximo con el primer elemento de la lista.
minimo = lista[0] # Inicializamos el mínimo con el primer elemento de la lista.
for i in range(1, len(lista)):
    if lista[i] > maximo: # Si encontramos un número mayor que el máximo actual, lo actualizamos.
        maximo = lista[i]
    if lista[i] < minimo: # Si encontramos un número menor que el mínimo actual, lo actualizamos.
        minimo = lista[i]
    lista_ordenada += [maximo] # Agregamos el máximo encontrado a la lista ordenada.
print(lista_ordenada)
````

#### Ejercicio 3. Selección de elementos pares e impares de una lista.

Sin usar la función `filter()`, escriba un algoritmo para obtener una lista con los elementos pares y otra con los elementos impares de una lista de números.

*Nota:* Este ejercicio puede resolverse mediante un recorrido secuencial y una evaluación condicional del residuo de cada número al dividirlo entre `2`.

````python
lista = [32, 11, 24, 1, -23, 4000, 32]
lista_pares = [] # Inicializamos la lista de pares vacía.
lista_impares = [] # Inicializamos la lista de impares vacía.
for num in lista:
    if num % 2 == 0: # Si el número es par, lo agregamos a la lista de pares.
        lista_pares += [num]
    else:
        lista_impares+=[num] # Si el número es impar, lo agregamos a la lista de impares.
print(lista_pares)
print(lista_impares)
````

### Ejercicio 4. Despliegue de un triángulo de asteriscos.

Escriba un algoritmo para desplegar un triángulo de asteriscos con una altura dada por el usuario. Por ejemplo, si el usuario ingresa `5`, el triángulo debería verse así:

```
    *
   ***
  *****
 *******
*********
```

*Nota:* Este ejercicio puede resolverse construyendo cada renglón a partir de una cantidad variable de espacios y asteriscos en cada iteración.

````python
n = input("Ingrese un número (máximo 10): ")
if int(n) > 10: # Verificamos que el número ingresado no sea mayor a 10.
    n = 10
lista = []
for i in range(1, int(n) + 1):
    espacios = " " * (int(n) - i) # Creamos una cadena de espacios que se repite (n - i) veces. 
    izquierda = "*" * i # Creamos una cadena de asteriscos que se repite i veces.
    print(espacios + izquierda + izquierda[:-1]) # Imprimimos la combinación de espacios y asteriscos para formar el triángulo.
````

### Ejercicio 5. Validación de ingreso de un número positivo.

Escriba un algoritmo que solicite al usuario ingresar un número positivo no mayor a `10`. Si el usuario ingresa un número negativo o mayor a `10`, el algoritmo debe mostrar un mensaje de error y solicitar nuevamente el ingreso hasta que se cumpla la condición usando un bucle `while` con `continue` y `break`.

*Nota:* Este ejercicio puede resolverse mediante un ciclo infinito controlado por validaciones sucesivas que interrumpen o reinician la ejecución según el valor ingresado.

````python
while True:
    entrada = input("Ingrese un número positivo (máximo 10): ")
    n = int(entrada)
    if n > 10: # Verificamos que el número ingresado no sea mayor a 10.
        print("Número demasiado grande, por favor ingrese un número menor o igual a 10.")
        continue
    elif n < 1: # Verificamos que el número ingresado no sea menor a 1.
        print("Número demasiado pequeño, por favor ingrese un número mayor o igual a 1.")
        continue
    else:
        break
print("Número válido.")
````

### Ejercicio 6. Triángulo de Pascal.

Sin usar la función `math.comb()`, escriba un algoritmo para generar el [triángulo de Pascal](https://es.wikipedia.org/wiki/Tri%C3%A1ngulo_de_Pascal) hasta una cantidad de filas dada por el usuario, con un máximo de `10` filas.

El triángulo de Pascal es un arreglo triangular de números donde cada número es la suma de los dos números directamente encima de él en la fila anterior. La primera fila del triángulo es `[1]`, la segunda fila es `[1, 1]`, la tercera fila es `[1, 2, 1]`, y así sucesivamente.

*Nota:* Este ejercicio puede resolverse construyendo cada fila a partir de la anterior, sumando pares consecutivos de elementos y agregando `1` en ambos extremos.

````python
n = 10
fila_inferior = [1, 1]
print([1])
for i in range(2, n + 1):
    fila_superior = [1] # La fila superior siempre comienza con 1.
    for j in range(1, i):
        fila_superior += [fila_inferior[j-1] + fila_inferior[j]] # Cada elemento de la fila superior es la suma de los dos elementos anteriores de la fila inferior.
    fila_superior += [1] # La fila superior siempre termina con 1.
    print(fila_superior)
    fila_inferior = fila_superior # La fila inferior se actualiza para la siguiente iteración.
````

### Ejercicio 7. Búsqueda del segundo valor más grande de una lista.

Sin usar las funciones `sort()`, `sorted()`, `max()` ni `min()`, escriba un algoritmo para obtener el segundo valor más grande de una lista de números.

*Nota:* Este ejercicio puede resolverse mediante un recorrido lineal con actualización de máximos, manteniendo en todo momento el valor mayor y el segundo mayor encontrados hasta ese momento.

````python
lista = [32, 11, 24, 1, -23, 4000, 32, 250]
mayor = lista[0]
segundo = lista[0]

for num in lista[1:]:
    if num > mayor:
        segundo = mayor
        mayor = num
    elif num > segundo and num != mayor:
        segundo = num

print(segundo)
````

### Ejercicio 8. Verificación de un palíndromo.

Sin usar métodos de cadenas ni rebanadas, escriba un algoritmo para determinar si una palabra es un palíndromo. Considere que la palabra está escrita en minúsculas y sin espacios.

*Nota:* Una estrategia simple consiste en construir manualmente la palabra invertida recorriendo sus caracteres y compararla al final con la palabra original.

````python
palabra = "reconocer"
invertida = ""

for caracter in palabra:
    invertida = caracter + invertida

if palabra == invertida:
    print("La palabra es un palíndromo.")
else:
    print("La palabra no es un palíndromo.")
````

### Ejercicio 9. Eliminación de elementos repetidos conservando el orden.

Sin usar tipos `set`, métodos de listas ni funciones predefinidas para eliminar duplicados, escriba un algoritmo para generar una nueva lista sin elementos repetidos y conservando el orden de aparición original.

*Nota:* Este ejercicio puede resolverse mediante una búsqueda secuencial: antes de agregar cada elemento a la nueva lista, se verifica si ya apareció anteriormente.

````python
lista = [3, 1, 3, 5, 1, 7, 5, 9]
lista_sin_repetidos = []

for num in lista:
    repetido = False
    for elemento in lista_sin_repetidos:
        if num == elemento:
            repetido = True
            break
    if not repetido:
        lista_sin_repetidos += [num]

print(lista_sin_repetidos)
````

### Ejercicio 10. Búsqueda del primer elemento repetido.

Sin usar tipos `set`, métodos ni funciones predefinidas para conteo, escriba un algoritmo para encontrar el primer valor que se repite en una lista. Si no existe ninguno, despliegue `None`.

*Nota:* Una forma directa de resolverlo es comparar cada elemento con los que están a su derecha mediante ciclos anidados y detener la búsqueda al encontrar la primera repetición.

````python
lista = [4, 7, 1, 9, 7, 3, 1]
repetido = None

for i in range(len(lista)):
    for j in range(i + 1, len(lista)):
        if lista[i] == lista[j]:
            repetido = lista[i]
            break
    if repetido != None:
        break

print(repetido)
````

### Ejercicio 11. Mezcla de dos listas ordenadas.

Sin usar `sort()`, `sorted()` ni métodos de listas, escriba un algoritmo que reciba dos listas ordenadas de menor a mayor y genere una nueva lista también ordenada con todos sus elementos.

*Nota:* Este problema suele resolverse con una mezcla ordenada, comparando los primeros elementos pendientes de cada lista y agregando siempre el menor a la lista resultado.

````python
lista_1 = [1, 4, 6, 10]
lista_2 = [2, 3, 7, 9, 12]
lista_mezclada = []
i = 0
j = 0

while i < len(lista_1) and j < len(lista_2):
    if lista_1[i] <= lista_2[j]:
        lista_mezclada += [lista_1[i]]
        i += 1
    else:
        lista_mezclada += [lista_2[j]]
        j += 1

while i < len(lista_1):
    lista_mezclada += [lista_1[i]]
    i += 1

while j < len(lista_2):
    lista_mezclada += [lista_2[j]]
    j += 1

print(lista_mezclada)
````
