#### Ejercicio 1. Consultar ventas diarias con una función.

Una tienda registró sus ventas de lunes a viernes en la lista `[120, 95, 140, 110, 160]`. Defina una función que recorra la colección y regrese la venta del primer día, la del último y la del penúltimo, validando además que existan suficientes registros.

```python
# La función recorre la lista con while y selecciona las posiciones requeridas.
def resumen_ventas(ventas):
    if len(ventas) < 2:
        return 'Se requieren al menos dos registros.'

    indice = 0
    primera = None
    penultima = None
    ultima = None

    while indice < len(ventas):
        if indice == 0:
            primera = ventas[indice]
        elif indice == len(ventas) - 2:
            penultima = ventas[indice]
        elif indice == len(ventas) - 1:
            ultima = ventas[indice]
        indice += 1

    return primera, ultima, penultima

resumen_ventas([120, 95, 140, 110, 160])
```

#### Ejercicio 2. Reorganizar bloques de una agenda con una función.

La agenda de una mañana está representada por `['correo', 'reunion', 'analisis', 'pruebas', 'reporte', 'cierre']`. Defina una función que genere tres vistas útiles de esa agenda: un bloque central, otro con actividades alternadas y otro en orden inverso.

```python
# El recorrido con for permite construir nuevas listas a partir de reglas simples.
def bloques_agenda(agenda):
    if len(agenda) < 4:
        return 'La agenda es demasiado corta.'

    centrales = []
    alternadas = []
    inversa = []

    for indice, actividad in enumerate(agenda):
        if 1 <= indice <= 3:
            centrales.append(actividad)
        if indice % 2 == 0:
            alternadas.append(actividad)

    for indice in range(len(agenda) - 1, -1, -1):
        inversa.append(agenda[indice])

    return centrales, alternadas, inversa

bloques_agenda(['correo', 'reunion', 'analisis', 'pruebas', 'reporte', 'cierre'])
```

#### Ejercicio 3. Actualizar un carrito de compra con validaciones.

Defina una función que reciba una lista con productos de un carrito y simule cambios antes de pagar: añadir un artículo, insertar otro en una posición intermedia y retirar el último, evitando dejar el carrito vacío.

```python
# La función modifica la lista y valida que siempre quede al menos un producto.
def actualizar_carrito(carrito):
    carrito.append('regla')
    carrito.insert(1, 'borrador')

    if len(carrito) > 1:
        carrito.pop()

    if not carrito:
        return 'El carrito quedó vacío.'

    return carrito

actualizar_carrito(['cuaderno', 'pluma'])
```

#### Ejercicio 4. Desempaquetar la ubicación de un dron.

El sistema de rastreo devolvió la tupla `(19.43, -99.13, 2240)`, correspondiente a latitud, longitud y altitud. Defina una función que desempaque esos datos y use `if` para clasificar la altitud.

```python
# La función desempaqueta la tupla y clasifica la altitud con una condición.
def describir_ubicacion(coordenadas):
    latitud, longitud, altitud = coordenadas

    if altitud >= 2000:
        nivel = 'alta'
    else:
        nivel = 'media o baja'

    return f'Latitud: {latitud}, Longitud: {longitud}, Altitud: {altitud} m ({nivel})'

describir_ubicacion((19.43, -99.13, 2240))
```

#### Ejercicio 5. Corregir el estado de un pedido.

Partiendo de la tupla `('pendiente', 'en ruta', 'entregado')`, defina una función que la convierta a lista, recorra los elementos con `while`, reemplace `"en ruta"` por `"en reparto"` y regrese el resultado como tupla.

```python
# La tupla se convierte a lista para poder modificar su contenido.
def corregir_estados(estados):
    copia = list(estados)
    indice = 0

    while indice < len(copia):
        if copia[indice] == 'en ruta':
            copia[indice] = 'en reparto'
        indice += 1

    return tuple(copia)

corregir_estados(('pendiente', 'en ruta', 'entregado'))
```

#### Ejercicio 6. Normalizar el título de una publicación.

A partir de la cadena `"  guia de python para entrevistas tecnicas  "`, defina una función que elimine los espacios sobrantes y convierta el texto a formato de título. Si la cadena queda vacía, debe regresar un mensaje de error.

```python
# strip() limpia extremos y una condición evita procesar cadenas vacías.
def normalizar_titulo(titulo):
    titulo = titulo.strip()

    if not titulo:
        return 'No hay contenido para normalizar.'

    return titulo.title()

normalizar_titulo('  guia de python para entrevistas tecnicas  ')
```

#### Ejercicio 7. Analizar un asunto de correo.

Con la cadena `"alerta: pago pendiente"`, defina una función que obtenga cuántas veces aparece la letra `a`, en qué posición comienza la palabra `pago` y si el asunto conserva el prefijo y sufijo esperados.

```python
# El recorrido del texto permite contar caracteres y una condición detecta una subcadena.
def analizar_asunto(asunto):
    contador_a = 0
    posicion_pago = -1

    for caracter in asunto:
        if caracter == 'a':
            contador_a += 1

    indice = 0
    while indice <= len(asunto) - len('pago'):
        if asunto[indice:indice + 4] == 'pago':
            posicion_pago = indice
            break
        indice += 1

    return {
        'cantidad_a': contador_a,
        'posicion_pago': posicion_pago,
        'inicia_alerta': asunto.startswith('alerta'),
        'termina_pendiente': asunto.endswith('pendiente')
    }

analizar_asunto('alerta: pago pendiente')
```

#### Ejercicio 8. Preparar una lista de invitados.

Transforme la cadena `"ana, luis, carla"` en una lista de nombres sin espacios extra y después únalos con el separador `" | "`. Defina una función que use un ciclo para limpiar cada nombre antes de unirlos.

```python
# La función recorre cada fragmento y conserva solo los nombres con contenido.
def preparar_invitados(datos):
    invitados = []

    for fragmento in datos.split(','):
        nombre = fragmento.strip().title()
        if nombre:
            invitados.append(nombre)

    return invitados, ' | '.join(invitados)

preparar_invitados('ana, luis, carla')
```

#### Ejercicio 9. Construir una ruta y un mensaje de bitácora.

#### Ejercicio 10. Crear un mensaje con una *f-string* y una condición.

Use las variables `nombre = 'Lucia'` y `tareas = 3` para construir una función que genere una oración mediante una *f-string*. La función debe usar `if` para distinguir entre singular y plural.

#### Ejercicio 11. Formatear números con una *f-string* dentro de una función.

Calcule el total con IVA de un precio de `149.9` y muéstrelo con dos decimales y símbolo de moneda. La función debe usar `if` para rechazar cantidades negativas.

```python
# La función valida el dato antes de calcular el total.
def total_con_iva(precio, iva=0.16):
    if precio < 0:
        return 'El precio no puede ser negativo.'

    total = precio * (1 + iva)
    return f'Total con IVA: ${total:.2f}'

total_con_iva(149.9)
```

#### Ejercicio 12. Alinear datos con una *f-string* y una función.

Despliegue una fila de texto tabular con el producto `Cuaderno`, la cantidad `7` y el costo `18.5`, alineando cada dato dentro de un ancho fijo. Use `if` para validar que la cantidad sea positiva.

```python
# La función produce una fila tabular solo si los datos son validos.
def fila_producto(producto, cantidad, costo):
    if cantidad <= 0:
        return 'La cantidad debe ser positiva.'

    return f'|{producto:<12}|{cantidad:^8}|${costo:>7.2f}|'

fila_producto('Cuaderno', 7, 18.5)
```

### Bloque 4. bytes y bytearray.

Los siguientes ejercicios retoman la notebook 19: codificación, decodificación e inspección y modificación de datos binarios.

#### Ejercicio 13. Codificar y decodificar texto con una función.

Convierta la cadena `"cafe"` a `bytes` usando `UTF-8` y luego recupere la cadena original. Defina una función que use un ciclo para inspeccionar los valores numéricos de los bytes generados.

```python
# El recorrido de bytes permite observar el valor entero de cada elemento.
def codificar_y_leer(texto):
    contenido = texto.encode('utf-8')
    valores = []

    for byte in contenido:
        valores.append(byte)

    return contenido, valores, contenido.decode('utf-8')

codificar_y_leer('cafe')
```

#### Ejercicio 14. Modificar un objeto `bytearray` con una función.

Cree una función que reciba `b'hola'`, cambie la primera letra por una mayúscula y agregue un signo de admiración al final. Use `if` para verificar que el contenido no esté vacío.

```python
# bytearray es mutable, por lo que permite actualizar bytes individuales.
def ajustar_saludo(datos_iniciales):
    datos = bytearray(datos_iniciales)

    if not datos:
        return 'No hay datos para modificar.'

    datos[0] = ord('H')
    datos.extend(b'!')
    return datos

ajustar_saludo(b'hola')
```

#### Ejercicio 15. Consultar valores de un diccionario mediante una función.

Defina una función que reciba un diccionario con los datos de un alumno y obtenga el nombre mediante indexación por clave y el turno mediante `get()` con un valor por defecto. Use `if` para validar que exista la clave `nombre`.

```python
# La función combina acceso directo, get() y una validación previa.
def datos_alumno(alumno):
    if 'nombre' not in alumno:
        return 'Falta la clave nombre.'

    return alumno['nombre'], alumno.get('turno', 'matutino')

datos_alumno({'nombre': 'Elena', 'grupo': 'A1', 'promedio': 9.4})
```

#### Ejercicio 16. Actualizar y eliminar datos de un diccionario.

A partir de un inventario inicial, defina una función que aplique cambios de existencias, agregue productos nuevos y elimine uno ya retirado del catálogo, validando antes que la clave correspondiente exista.

```python
# update() fusiona cambios y pop() elimina una clave devolviendo su valor.
def actualizar_inventario(inventario):
    inventario.update({'lapices': 15, 'reglas': 3})

    if 'gomas' in inventario:
        inventario.pop('gomas')

    return inventario

actualizar_inventario({'lapices': 12, 'gomas': 5})
```

#### Ejercicio 17. Recorrer un diccionario y formatear sus datos.

Con el diccionario de precios dado, construya una lista de etiquetas de texto usando `items()` y un ciclo `for`. Use `if` para omitir precios no positivos.

```python
# items() entrega pares clave-valor que pueden recorrerse directamente.
def etiquetar_precios(precios):
    etiquetas = []

    for articulo, precio in precios.items():
        if precio > 0:
            etiquetas.append(f'{articulo}: ${precio}')

    return etiquetas

etiquetar_precios({'cuaderno': 45, 'pluma': 12, 'borrador': 8})
```

### Bloque 6. Conjuntos y frozenset.

#### Ejercicio 18. Eliminar duplicados con un conjunto.

Con la lista `['Ana', 'Luis', 'Ana', 'Marta', 'Luis']`, defina una función que cree un conjunto de nombres unicos y evalúe si `Ana` pertenece a ese conjunto. Use un ciclo para construir primero una lista depurada.

```python
# Un recorrido previo permite decidir qué elementos conservar antes de crear el conjunto.
def participantes_unicos(inscritos):
    depurados = []

    for nombre in inscritos:
        if nombre not in depurados:
            depurados.append(nombre)

    participantes = set(depurados)
    return participantes, 'Ana' in participantes

participantes_unicos(['Ana', 'Luis', 'Ana', 'Marta', 'Luis'])
```

#### Ejercicio 19. Comparar conjuntos con operaciones básicas.

Dados los conjuntos de estudiantes `python` y `sql`, defina una función que obtenga la unión, la intersección y la diferencia de quienes solo están en `python`. Use `if` para verificar que ambos conjuntos tengan elementos.

```python
# Las operaciones entre conjuntos resumen relaciones frecuentes entre colecciones de datos.
def comparar_grupos(python, sql):
    if not python or not sql:
        return 'Ambos conjuntos deben tener elementos.'

    return python | sql, python & sql, python - sql

comparar_grupos({'Ana', 'Luis', 'Elena'}, {'Luis', 'Mario', 'Elena'})
```

#### Ejercicio 20. Usar un `frozenset` como clave.

Defina una función que modele distancias entre ciudades usando claves no ordenadas y consulte la ruta entre `CDMX` y `Puebla` aunque el orden de entrada cambie. Si la ruta no existe, debe reportarlo.

```python
# frozenset es hashable y sirve como clave cuando el orden de los elementos no importa.
def buscar_ruta(origen, destino):
    rutas = {
        frozenset({'CDMX', 'Puebla'}): 140,
        frozenset({'CDMX', 'Toluca'}): 70,
    }
    clave = frozenset({origen, destino})

    if clave in rutas:
        return rutas[clave]

    return 'La ruta no está registrada.'

buscar_ruta('Puebla', 'CDMX')
```
