# Declaraciones y bloques de código.

## Flujo de ejecución del  código.

El intérprete de *Python* es capaz de leer, evaluar y ejecutar una sucesión de instrucciones línea por línea de principio a fin. A esto se le conoce como flujo de ejecución de código.

Los lenguajes de programacion modernos pueden ejecutar porciones de código dependiendo de ciertas condiciones. Estas porciones de código también se conocen como "bloques" y deben de ser delimitados sintácticamente. 

## Declaraciones.

Las declaraciones (statements) son expresiones capaces de contener a un bloque de código, las cuales se ejecutarán e incluso repetirán en caso de que se cumplan ciertas condiciones, dependiendo del tipo de declaración.

### Delimitación de expresiones y declaraciones en otros lenguajes.

Algunos lenguajes de programación identifican el final de una expresión mendiante el uso del punto y coma ```;``` y también suelen delimitar bloques de código encerrándolos entre llaves ```{``` ```}```.

**Ejemplo:**

* El siguiente código ilusta el uso de llaves y los finales con punto y coma en *Javascript*.
  * Se despegará una sucesión de números de `1` al `10`usando la declaración `for`.
  * Se desplegará el mensaje `Esta es la línea final.`.


```javascript

for (let i = 1; i <= 10; i++) {
    console.log(i);
}
console.log("Esta es la línea final.");
```

### Indentación en otros lenguajes de programación.

En otros lenguajes de programación, la indentación (dejar espacios o tabuladores al principio de en una línea) se considera como una buena práctica y una regla de estilo a fin de hacer más legible el código. Sin embargo, la indentación es opcional.

**Ejemplo:**

* El siguiente código de *Javascript* es idéntico al anterior. El intérprete ignora los retornes de línea y las indentaciones.

```javascript
for (let i = 1; i <= 10; i++) {console.log(i);} console.log("Esta es la línea final.");
```








## Declaraciones en *Python*.

En el caso de *Python* la indentación es un elemento sintáctico que delimita los bloques de código en una declaración. El *PEP-8* indica que la indentación correcta es usando cuatro espacios.

```
<flujo>
...
...
<declaración>:
    <código>
<flujo>
...
```

Donde:
* `<flujo>` representa el flujo de código que contiene múltiples expersiones que conforman el bloque de código principal.
* `<código>` es un bloque de código delimitado mediante indentación.
* `<declaración>` es una instrucción que afecta el comportamiento del bloque `<código>`.

**Ejemplos:**

* La siguiente celda ejemplifica del uso de indentación en *Python* para delimitar bloques de código.
  * Se despegará una sucesión de números de `1` al `10`usando la declaración `for ...in`
  * Se desplegará el mensaje `Esta es la línea final.`.

````python
for i in range(1, 11):
    print(i)
print('Esta es la línea final.')
````

* La siguiente celda indentación línea final, haciendo que el mensaje `Esta es la línea final.` se imprima en cada iteración.

````python
for i in range(1, 11):
    print(i)
print('Esta es la línea final.')
````

### Declaraciones anidadas.

Es muy común que el código incluya declaraciones dentro de otras declaraciones. En ese caso, sólo es necesario ajustar la indentación a las declaraciones andidadas.

```
<flujo>
<declaración>:
    ....
    <declaración anidada>:
        ....
<flujo>
```

**Ejemplo:**

* La siguiente celda incluye una estructura de declaraciones condicionales anidadas en una delaración de iteración.

````python
'''Esta celda realizará una iteración de números en el rango 
entre el ```1``` y ```10``` y mostrará un mensaje dependiendo
si el número es par o non.'''

for i in range (1, 11):
    if i % 2 == 0:
        valor = 'par'
    else:
        valor = 'non'
    print(f'El número {i} es {valor}.')
    
print('Esta es la línea final.')
````

* La siguiente celda sustituye las declaraciones condicionales con una expresión ternaria.

````python
'''Esta celda realizará una iteración de números en el rango 
entre el ```1``` y ```10``` y mostrará un mensaje dependiendo
si el número es par o non.'''

for i in range (1, 11):
    print(f'El número {i} es {"par" if i % 2 == 0 else "non"}.')
print('Esta es la línea final.')
````
