# Introducción.

[*Python*](http://python.org) es un lenguaje de programación de alto nivel, interactivo e interpretado, creado por Guido Van Rossum en 1991. El nombre del lenguaje es en honor al programa cómico de la televisión británica llamado “Monty Python Flying Circus”.

Es de código abierto, multi-plataforma y se adecua a diversos paradigmas de programación (programación funcional, programación orientada a objetos, programación procedimental, etc.).

*Python* se ha vuelto muy popular ya sea como un primer lenguaje de programación o como el lenguaje seleccionado por defecto por diversos proyectos de primer nivel.

## Características de *Python*.

A lo largo de estos cursos se explorarán y aprovecharán las características que hacen de *Python* un lenguaje tan popular y poderoso.

* Sintaxis muy clara y legible.
* Fuerte capacidad de introspección.
* Orientación a objetos intuitiva.
* Expresión del código procedimental.
* Altamente modular, soporta paquetes jerárquicos.
* Enfocado en el uso de excepciones para el manejo de errores.
* Tipos de datos dinámicos de muy alto nivel.
* Extensa [biblioteca estándar](https://docs.python.org/3/library/) y módulos de terceros para prácticamente todas las tareas.
* Cuenta con un amplio catálogo de paquetes de software publicados en el repositorio [*pypi*](https://pypi.org).
* Es capaz de hacer llamadas a bibliotecas de binarios escritas en lenguajes como *C*, *C++*, *Fortran*, etc.
* Puede ser integrado dentro de las aplicaciones o plataformas como una interfaz de *scripting*.

## *Python 2* y *Python 3*.

*Python 3*  es una versión revisada del lenguaje, la cual fue publicada en 2009 y que incluye modificaciones y mejoras que lo hacen incompatible con código de versiones previas; mientras que *Python 2* es una versión que es compatible con código antiguo. 

*Python 3* presenta mejoras notables con respecto a *Python 2*, sin embargo aún existe mucho código "heredado" compatible con *Python 2*.

La publicación y soporte de nuevas versiones de *Python 2* cesó a partir del año 2020.

**NOTA:** Este curso se basa en *Python 3*, pero se discutirá brevemente sobre las diferencias entre cada versión cuando se considere pertinente.

## Breve introducción a los lenguajes de programación.

Un lenguaje es un conjunto de cadenas de símbolos con los que se pueden crear mensajes. De ese modo los mensajes son transmitidos de un emisor a un receptor. Aún cuando en la naturaleza se pueden identificar ciertos lenguajes, los seres humanos hemos desarrollado lenguajes de diversos tipos y gran complejidad.

Los lenguajes constan principalmente de la gramática, la cual trata sobre la construcción del lenguaje, y la semántica, la cual trata sobre el significado del lenguaje.

A su vez, la gramática consta de: 

* Morfología: cómo se construyen las notaciones (género, tiempos, declinaciones).
* Sintaxis: cómo se deben escribir las notaciones (orden, estructura).

El cómputo moderno y sus lenguajes son el resultado de contribuciones de científicos y matemáticos a lo largo de varios siglos:

* [**George Boole**](https://en.wikipedia.org/wiki/George_Boole) (1815-1864) desarrolló el álgebra que lleva su nombre, un sistema lógico basado en valores de verdad (*True*/*False*) y operaciones como *AND*, *OR* y *NOT*. El álgebra de Boole es la base matemática de todos los circuitos digitales y de la lógica de programación.

* [**Ada Lovelace**](https://en.wikipedia.org/wiki/Ada_Lovelace) (1815-1852) describió el primer algoritmo pensado para ser ejecutado por una máquina —la máquina analítica de Charles Babbage— y es considerada la primera programadora de la historia.

* [**Kurt Gödel**](https://en.wikipedia.org/wiki/Kurt_G%C3%B6del) (1906-1978) demostró en 1931 sus célebres *Teoremas de Incompletitud*: en cualquier sistema formal suficientemente expresivo existen proposiciones verdaderas que no pueden ser demostradas dentro del propio sistema. Este resultado sacudió los fundamentos de las matemáticas y planteó por primera vez de forma rigurosa los límites de lo que puede calcularse o demostrarse, sentando las bases filosóficas de la teoría de la computación.

* [**Alan Turing**](https://en.wikipedia.org/wiki/Alan_Turing) (1912-1954) y [**Alonzo Church**](https://en.wikipedia.org/wiki/Alonzo_Church) (1903-1995), inspirados en parte por el trabajo de Gödel, formalizaron de forma independiente el concepto de *algoritmo* y los límites de la computación. Turing lo hizo mediante su modelo teórico de la *Máquina de Turing*; Church mediante el *Cálculo Lambda*, que es la base matemática de los lenguajes funcionales modernos. Ambos demostraron que existen problemas que ninguna máquina podrá resolver jamás, como el famoso *Problema de la Parada* (*Halting Problem*).

* [**Claude Shannon**](https://en.wikipedia.org/wiki/Claude_Shannon) (1916-2001) fundó la *Teoría de la Información* y demostró que el álgebra de Boole podía aplicarse directamente al diseño de circuitos eléctricos, haciendo posible la construcción de computadoras digitales.

* [**John von Neumann**](https://en.wikipedia.org/wiki/John_von_Neumann) (1903-1957) propuso la arquitectura de *programa almacenado*, en la que tanto los datos como las instrucciones residen en la misma memoria. Esta arquitectura es la base de prácticamente todos los procesadores modernos.

* [**Grace Hopper**](https://en.wikipedia.org/wiki/Grace_Hopper) (1906-1992) desarrolló el primer compilador e impulsó la idea de que los programas podían escribirse en un lenguaje cercano al humano y traducirse automáticamente a código máquina, democratizando así la programación.

* [**John Backus**](https://en.wikipedia.org/wiki/John_Backus) (1924-2007) lideró el desarrollo de *FORTRAN* en 1957, el primer lenguaje de alto nivel de uso extendido, y posteriormente creó la notación *BNF* (*Backus-Naur Form*), que se convirtió en el estándar para describir la gramática de los lenguajes de programación.

Los lenguajes de programación actuales, a diferencia de los lenguajes humanos, tienen una morfología rígida y simplificada con el fin de ejecutar instrucciones específicas en los sistemas de cómputo.

### Lenguajes de alto y bajo nivel.

Los lenguajes de bajo nivel constan de un conjunto básico de instrucciones que son ejecutados directamente por la unidad de procesamiento de un sistema de cómputo, tal como es el caso del lenguaje ensamblador. Dichos lenguajes están ligados intrínsecamente al tipo de procesador que los ejecuta y resultan ser muy complicados de elaborar e interpretar por las personas. 

Por su parte, los lenguajes de alto nivel son más accesibles para el ser humano e incluso menos dependientes del tipo de hardware, pero deben de ser a su vez traducidos a lenguaje de bajo nivel.

### Lenguajes compilados e interpretados.

Los lenguajes de alto nivel interactúan con los sistemas de cómputo de dos formas.

* Mediante un compilador, el cual traduce el código de un programa a lenguaje de bajo nivel, dando por resultado un "archivo binario" el cual es susceptible de ser ejecutado.
* Mediante un intérprete, el cual ejecuta de inmediato las instrucciones que se ingresan.

Por lo general los lenguajes compilados son más rápidos y consumen menos recursos que los lenguajes interpretados en vista de que el archivo resultante es código de bajo nivel, mientras que los lenguajes interpretados deben seguir un proceso a través de varios niveles de abstracción hasta que las instrucciones son ejecutadas por el sistema.

***Python* es un lenguaje interpretado de alto nivel.**

## Entornos interactivos, *scripts* y *'Hola Mundo'*.

Debido a que *Python* es un lenguaje interpretado, es posible utilizarlo mediante un entorno interactivo (*shell*) o mediante el uso de guiones (*scripts*).

### El entorno interactivo.

El entorno interactivo (*shell*) de *Python 3* se ejecuta desde una terminal de texto de la siguiente manera:

```
python3
```

En algunos sistemas, el comando ```python``` también apunta a *Python 3*.
El entorno interactivo se verá de forma similar a la siguiente:

```
Python 3.12.3 (main, Mar  3 2026, 12:15:18) [GCC 13.3.0] on linux
Type "help", "copyright", "credits" or "license" for more information.
>>>
```

**NOTA:** Para salir del entorno interactivo se debe teclear ```exit()``` y oprimir la tecla <kbd>Intro</kbd>.

### *'Hola Mundo'* en el entorno interactivo.
Para desplegar el texto `Hola Mundo` desde el entorno interactivo sólo es necesario teclear lo siguiente:

```
>>> print("Hola, Mundo.")
```

Y el resultado es el siguiente:

```
Hola, Mundo.
>>>
```

### *'Hola Mundo'* en la *notebook* de *Jupyter*.

Las *notebooks* de *Jupyter* funcionan mediante *kernels* los cuales son conexiones a un intérprete del lenguaje correspondiente.

Para ejecutar un comando en una *notebook* de *Jupyter* con el *kernel* de *Python 3*, sólo es necesario escribir el código en una terminal de tipo ```Code``` y ejecutarla.

**Ejemplo:**

````python
print("Hola, Mundo.")
````

### Ejecución de *scripts*.

Los *scripts* de *Python* son archivos de texto que contienen código que será leído, evaluado y ejecutado por el intérprete de *Python*. Dichos archivos tienen la extensión ```.py```. Algunos editores de texto cuentan con reconocimiento de la sintaxis de *Python* y utilizan colores para identificar sus componentes.

#### El *script* ```src/holamundo.py```.

El documento  [```src/holamundo.py```](src/holamundo.py) contiene el siguiente código.

``` python
#!/usr/bin/env python                                                     
print("Hola, Mundo.")
```

#### Ejecución mediante el intérprete.

Para ejecutar el *script*, es necesario que el intérprete acceda al archivo con una sintaxis similar a: 

``` 
python3 <script>
```
 Donde:

 * `<script>` es un archivo de texto que contiene código escrito en *Python*.

 En algunos sistemas, ```python``` puede apuntar a *Python 3* y también funcionar.

 El intérprete de *Python* leerá, interpretará y ejecutará el código del *script* línea por línea de principio a fin.

**Ejemplo:**

Las *notebooks* de *Jupyter* permiten enviar comandos al sistema anteponiendo un signo de admiración ```!``` antes del comando.

* La siguiente celda ejecutará el *script* `src/holamundo.py` usando el intérprete de *Python 3*.

**Nota:** En la mayoría de entornos actuales el comando recomendado es ```python3```. Si en tu sistema ```python``` ya apunta a *Python 3*, ambos comandos funcionarán.

````python
!python3 src/holamundo.py
````

#### Ejecución de un *script* en *Jupyter*.

Jupyter puede ejecutar *scripts* mediante el "comando mágico" ```%run```.

Los comandos mágicos de *IPython* serán estudiados más adelante.

**Ejemplo:**

* La siguiente celda ejecutará el *script* que se encuentra en la ruta relativa `src/holamundo.py`.

````python
%run src/holamundo.py
````

#### Ejecución directa de un *script* en entornos basados en *UNIX*.

La ejecución directa del *script* en entornos basados en *UNIX* requiere que se indique la ruta en la que se encuentra el intérprete de *Python* mediante el texto: 

```
#! /usr/bin/env python 
```

A los símbolos `#!` se les conoce como *shebang* y le indican al sistema la ruta en la que se encuentra el intérprete (no necesariamente de *Python*) que ejecutará el *script*.

**Nota:** Es necesario que la ruta definida en el *shebang* apunte al ejecutable de un intérprete de *Python*.

Es necesario que el *script* cuente con los permisos de ejecución necesarios. Para asignar permisos de ejecución a un archivo se utiliza el comando: 

```
chmod +x <ruta>
```

Donde:

* `<ruta>` es la ruta donde se encuentra el *script*.

**Ejemplo:**

* En caso de que esta *notebook* se encuentre corriendo en un sistema *GNU/Linux* o *macOS*, el siguiente comando asignará permisos de ejecución al *script* localizado en ```src/holamundo.py```.

**Advertencia:** Las siguientes celdas no funcionarán en *Windows*.

````python
!chmod +x src/holamundo.py
````

* La siguiente celda se ejecutará si existe un intérprete en la ruta definida en el *shebang*.  

````python
!./src/holamundo.py
````

#### Ejecución en *Windows*.

En el caso de *Windows*, el sistema relacionará a los archivos con la extensión ```.py``` con el intérprete de *Python*, por lo que con hacer doble click en el archivo, éste abrirá una terminal, el *script* se ejecutará y tan pronto termine, cerrará la terminal.

Cuando se ejecute el *script* ```src/holamundo.py``` en *Windows*, se abrirá y cerrará una terminal de forma casi inmediata.

**Ejemplo:**

* La siguiente celda se ejecutará en un entorno basado en *Windows* si existe la ruta a un interprete de *Python* configurada en el sistema.

**Advertencia:** La siguiente celdas no funcionarán en entornos distintos a *Windows*.

````python
!src\holamundo.py
````

## Codificación de caracteres.

*Python 3* utiliza la codificación [*UTF-8*](https://www.utf8-chartable.de/unicode-utf8-table.pl) por defecto.

*Python 2* utilizaba la codificación [*ASCII*](http://www.asciitable.com/), por lo que desplegar caracteres especiales como la "ñ" y los acentos generaban un mensaje de error.

### Codificación *UTF-8* explícita.
Para indicarle al intérprete que utilice la codificación UTF-8 se debe incluir la siguiente línea al principio del *script*:
``` python
# -*- coding: utf-8 -*-
```
**Ejemplo:**

El siguiente código define un *script* que incluye la configuración de codificación *UTF-8*, la cual es opcional, pero recomendable.

``` python
#! /usr/bin/python
# -*- coding: utf-8 -*- 
print("Hola niños")
input()
```

## La comunidad de *Python*.

*Python* es un lenguaje que ha crecido gracias a una amplia y participativa comunidad. Anteriormente, Guido Van Rossum tenía la última palabra en el desarrollo de *Python* en su carácter de *Tirano Benevolente*, pero desde que se retiró del proyecto, la [*Python Software Foundation*](https://www.python.org/psf-landing/) es la encargada de la evolución y desarrollo del lenguaje y está compuesta por diversos miembros de la comunidad.

A los programadores de *Python* se les conoce como "pythonistas" y con el tiempo han desarrollado todo un compendio de mejores prácticas y códigos de estilo para programar en este lenguaje.

### El *Zen de Python*.

El *Zen de Python* es una lista de aforismos que indican de forma general lo que se espera del código hecho en *Python*, y ha sido reconocido de tal forma por la comunidad que puede ser consultado desde el intérprete de *Python* de la siguiente manera.

````python
import this
````

Aquí la traducción de los aforismos.

* Bello es mejor que feo.
* Explícito es mejor que implícito.
* Simple es mejor que complejo.
* Complejo es mejor que complicado.
* Plano es mejor que anidado.
* Escaso es mejor que denso.
* La legibilidad cuenta.
* Los casos especiales no son tan especiales como para romper las reglas.
* Aunque lo práctico le gana a la pureza.
* Los errores no deben pasar en silencio.
* A menos que sean silenciados explícitamente.
* En caso de ambigüedad, resiste la tentación a adivinar.
* Debería haber una --y preferiblemente sólo una-- manera obvia de hacer las cosas.
* Aunque esa manera no parezca ser tan obvia a menos que seas Holandés.
* Ahora es mejor que nunca.
* Aunque nunca es normalmente mejor que ahora "mismo".
* Si la implementación es difícil de explicar, es una mala idea.
* Si la implementación es fácil de explicar, podría ser una buena idea.
* Los Espacios de Nombres son una magnífica idea -- ¡Hagamos más de esos!

### Las *PEP* de *Python*.

Es muy probable que algún pythonista comparta alguna inquietud o tenga alguna propuesta relacionada con algún aspecto del lenguaje y es por eso que existe un espacio que compendia una inmensa cantidad de propuestas e ideas relacionadas con el lenguaje. 

[*PEP*](https://peps.python.org/) es el acrónimo de "Propuesta de mejora de *Python*" y es el sitio en el que se puede proponer, consultar e incluso debatir sobre todas estas inquietudes de forma abierta y directa con los desarrolladores del lenguaje.

Por ejemplo, la [*PEP-8*](https://peps.python.org/pep-0008/) es particularmente interesante, ya que se refiere a las reglas de estilo que se sugieren para programar en *Python*.

### El índice de paquetes de *Python* (*PyPI*).

*Python* es un lenguaje que viene con una gran biblioteca estándar, sin embargo existe una biblioteca en línea aún más grande y que sirve como repositorio de diversos proyectos publicados por sus desarrolladores para uso de la comunidad.
El sitio [https://pypi.org](https://pypi.org) no sólo enlista a dichos proyectos, sino que además cuenta con ```pip```, una herramienta que puede instalar de forma local los paquetes correspondientes a uno o diversos proyectos contenidos en *PyPI*.
