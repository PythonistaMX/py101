# Tipos ```bytes``` y ```bytearray```.

## El tipo ```bytes```.

Los objetos de tipo ```bytes``` son una colección ordenada, indexable e inmutable de valores enteros entre ```0``` y ```255```. Dependiendo de la codificación empleada (por ejemplo *ASCII* o *UTF-8*), estos valores pueden interpretarse como caracteres. Se definen anteponiendo la letra ```b``` a los apóstrofes o comillas.
```
b'<texto>'
b"<texto>" 
```

**Ejemplos:*** Las siguientes celdas crearán un objeto de tipo ```bytes```, mostrarán su tipo, permitirán acceder a sus elementos por índice y rango, e intentarán modificarlo para evidenciar su inmutabilidad.

````python
palabra = b"Hola"
````

### El método ```decode()```.

Este método funciona de manera inversa al método ```encode()``` de los objetos tipo ```str```. 
Se ingresan como argumentos el valor de codificación ligado al identificador _encoding_ y el valor de manejo de errores de codificación ligado al identificador _errors_.



**Ejemplo:*** La siguiente celda regresará un objeto de tipo ```str``` al decodificar un objeto ```bytes``` usando la codificación indicada.

````python
b'Hola,'.decode(encoding="utf-8")
````

### El método ```hex()```.

Regresa un objeto tipo ```str``` que contiene la representación en hexadecimal de cada elemento del objeto tipo ```bytes```.

**Ejemplo:*** La siguiente celda regresará la representación hexadecimal del objeto ```bytes```.

````python
b'Hola'.hex()
````

### El método ```fromhex()```.

Regresa un objeto tipo ```bytes``` a partir de un objeto ```str``` que contiene la representación hexadecimal de uno o más bytes.

**Ejemplo:*** La siguiente celda regresará un objeto de tipo ```bytes``` a partir de una cadena hexadecimal.

````python
bytes.fromhex('486f6c61')
````

## Los objetos de tipo ```bytearray```.

Los objetos de tipo ```bytearray``` son una colección ordenada, indexable y mutable de valores enteros entre ```0``` y ```255```. Dependiendo de la codificación empleada (por ejemplo *ASCII* o *UTF-8*), estos valores pueden interpretarse como caracteres.

**Ejemplos:*** Las siguientes celdas crearán un objeto ```bytearray```, consultarán elementos por índice, modificarán valores y eliminarán elementos para evidenciar su mutabilidad.

````python
arreglo = bytearray(b'hola')
````

````python
arreglo
````

````python
arreglo[3]
````

````python
arreglo[3] = 101
````

````python
arreglo
````

````python
del arreglo[0]
````

````python
arreglo
````

````python
dir(arreglo)
````

### El método ```clear()```.

**Ejemplo:*** La siguiente celda vaciará el objeto ```bytearray``` y regresará su estado final.

````python
arreglo = bytearray(b'hola')
````

````python
arreglo.clear()
````

````python
arreglo
````

### El método ```reverse()```.

**Ejemplo:*** La siguiente celda invertirá el orden de los elementos del ```bytearray``` y regresará el resultado.

````python
arreglo = bytearray(b'hola')
````

````python
arreglo.reverse()
````

````python
arreglo
````

### El método ```append()```.

**Ejemplo:*** La siguiente celda añadirá un byte al final del ```bytearray``` y regresará el arreglo actualizado.

### El método ```insert()```.

### El método ```remove()```.

````python
arreglo.remove(108)
````

````python
arreglo
````

### El método ```copy()```.

**Ejemplo:*** La siguiente celda creará una copia del ```bytearray``` y comparará igualdad y referencia entre ambos objetos.

````python
arreglo = bytearray(b'hola')
````

````python
nuevo_arreglo = arreglo.copy()
````

````python
nuevo_arreglo == arreglo
````

````python
nuevo_arreglo
````

````python
nuevo_arreglo is arreglo
````
