# Tipos ```bytes``` y ```bytearray```.

## El tipo ```bytes```.

Los objetos de tipo ```bytes``` son una colección ordenada, indexable e inmutable de valores enteros entre ```0``` y ```255```. Dependiendo de la codificación empleada (por ejemplo *ASCII* o *UTF-8*), estos valores pueden interpretarse como caracteres. Se definen anteponiendo la letra ```b``` a los apóstrofes o comillas.
```
b'<texto>'
b"<texto>" 
```

## Métodos compartidos por los objetos ```str```, ```bytes``` y ```bytearray```.

Los objetos de tipo ```bytes``` y ```bytearray``` comparten con los de tipo ```str``` los siguientes métodos de texto (lista validada para *Python 3*):

De forma breve:
* Métodos de transformación y formato: ```capitalize()```, ```lower()```, ```upper()```, ```title()```, ```swapcase()```, ```center()```, ```ljust()```, ```rjust()```, ```zfill()```.
* Métodos de limpieza y separación: ```strip()```, ```lstrip()```, ```rstrip()```, ```split()```, ```rsplit()```, ```splitlines()```, ```partition()```, ```rpartition()```, ```join()```.
* Métodos de búsqueda y reemplazo: ```find()```, ```rfind()```, ```index()```, ```rindex()```, ```count()```, ```replace()```, ```startswith()```, ```endswith()```.
* Métodos de validación: ```isalnum()```, ```isalpha()```, ```isascii()```, ```isdigit()```, ```islower()```, ```isspace()```, ```istitle()```, ```isupper()```.
* Métodos de mapeo/normalización: ```maketrans()```, ```translate()```, ```removeprefix()```, ```removesuffix()```.

**Ejemplos:**

* Las siguientes celdas crearán un objeto de tipo ```bytes```, mostrarán su tipo, permitirán acceder a sus elementos por índice y rango, e intentarán modificarlo para evidenciar su inmutabilidad.

````python
palabra = b"Hola"
````

* La siguiente celda ejecutará la función ```type()``` con el objeto ```palabra```, mostrando que es de tipo ```bytes```.

````python
type(palabra)
````

* La siguiente celda ejecutará la función ```print()``` con el objeto ```palabra```, desplegando su representación en pantalla.

````python
print(palabra)
````

* La siguiente celda desplegará el elemento en el índice ```0``` de ```palabra```, el cual es el valor numérico del byte correspondiente al carácter ```'H'```.

````python
palabra[0]
````

* La siguiente celda desplegará el rebanado ```palabra[2:4]```, obteniendo un objeto ```bytes``` con los elementos en las posiciones ```2``` y ```3```.

````python
palabra[2:4]
````

* La siguiente celda intentará modificar el elemento ```palabra[1]```, lo que desencadenará un error de tipo ```TypeError``` por la inmutabilidad de los objetos ```bytes```.

````python
palabra[1] = b'm'
````

* La siguiente celda ejecutará la función ```str()``` sin especificar codificación, obteniendo la representación literal del objeto ```bytes``` como cadena.

````python
str(b'Hola')
````

* La siguiente celda ejecutará la función ```str()``` especificando la codificación ```"ascii"```, obteniendo el objeto ```str``` correspondiente.

````python
str(b'Hola', encoding="ascii")
````

### El método ```decode()```.

Este método funciona de manera inversa al método ```encode()``` de los objetos tipo ```str```. 
Se ingresan como argumentos el valor de codificación ligado al identificador _encoding_ y el valor de manejo de errores de codificación ligado al identificador _errors_.



**Ejemplo:**

* La siguiente celda regresará un objeto de tipo ```str``` al decodificar un objeto ```bytes``` usando la codificación indicada.

````python
b'Hola,'.decode(encoding="utf-8")
````

### El método ```hex()```.

Regresa un objeto tipo ```str``` que contiene la representación en hexadecimal de cada elemento del objeto tipo ```bytes```.

**Ejemplo:**

* La siguiente celda regresará la representación hexadecimal del objeto ```bytes```.

````python
b'Hola'.hex()
````

### El método ```fromhex()```.

Regresa un objeto tipo ```bytes``` a partir de un objeto ```str``` que contiene la representación hexadecimal de uno o más bytes.

**Ejemplo:**

* La siguiente celda regresará un objeto de tipo ```bytes``` a partir de una cadena hexadecimal.

````python
bytes.fromhex('486f6c61')
````

## Los objetos de tipo ```bytearray```.

Los objetos de tipo ```bytearray``` son una colección ordenada, indexable y mutable de valores enteros entre ```0``` y ```255```. Dependiendo de la codificación empleada (por ejemplo *ASCII* o *UTF-8*), estos valores pueden interpretarse como caracteres.

**Ejemplos:**

* Las siguientes celdas crearán un objeto ```bytearray```, consultarán elementos por índice, modificarán valores y eliminarán elementos para evidenciar su mutabilidad.

````python
arreglo = bytearray(b'hola')
````

* La siguiente celda desplegará el contenido del objeto ```arreglo```.

````python
arreglo
````

* La siguiente celda desplegará el valor numérico del byte en el índice ```3``` de ```arreglo```.

````python
arreglo[3]
````

* La siguiente celda sustituirá el elemento en el índice ```3``` de ```arreglo``` por el valor numérico ```101```, correspondiente al carácter ```'e'```.

````python
arreglo[3] = 101
````

* La siguiente celda desplegará el contenido actualizado de ```arreglo``` tras la sustitución.

````python
arreglo
````

* La siguiente celda eliminará el elemento en el índice ```0``` de ```arreglo```.

````python
del arreglo[0]
````

* La siguiente celda desplegará el contenido de ```arreglo``` tras la eliminación del primer elemento.

````python
arreglo
````

* La siguiente celda ejecutará la función ```dir()``` sobre ```arreglo```, listando todos sus atributos y métodos disponibles.

````python
dir(arreglo)
````
