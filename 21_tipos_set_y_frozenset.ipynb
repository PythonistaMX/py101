# Objetos tipo ```set``` y ```frozenset```.

Existen colecciones de objetos en *Python* que permiten realizar operaciones con conjuntos. Son una colección de objetos de valor único y no ordenados.
Estos tipos de datos no son indexables y por lo tanto tampoco se pueden rebanar.

## Objetos tipo ```set```.

### Definición de un objeto tipo ```set```.

Los elementos contenidos en un objeto tipo _set_ deben ser *hashables*.

Los objetos tipo _set_ se definen mediante el uso de llaves ```{```, ```}```, pero a diferencia de los objetos de tipo ```dict```, sólo se ingresa una sucesión de objetos *hashables*.

```
{<valor 1>, <valor 2>, ..., <valor n>}
```
Donde:
- ```<valor 1>```, ```<valor 2>```, ..., ```<valor n>``` son objetos *hashables*.

La función ```set()``` permite crear conjuntos a partir de un objeto *iterable*. En ese caso, cada elemento producido por el *iterable* también debe ser *hashable*.

``` 
set(<objeto iterable>)
```
**Ejemplos:** 

````python
{1, 'dos', 3.0, '4'}
````

* La siguiente celda define un objeto ```set``` con valores numéricos equivalentes y cadenas distintas, mostrando que los duplicados se eliminan automáticamente.

````python
{2, 2., 1 + 1, "dos", '2'}
````

* La siguiente celda ejecutará la función ```set()``` con un objeto ```dict```, obteniendo un conjunto con únicamente las claves del diccionario.

````python
set({'id_1': 3, 'id_2': 4, 'id_3': 5, 'id_4': 6})
````

````python
{[[12, 3, 27], [True, 60]]}
````

````python
set([[12, 3, 27], [True, 60]])
````

* La siguiente celda define un objeto ```set``` con elementos de tipo ```tuple```. En este caso las tuplas son *hashables* porque todos sus elementos también lo son.

````python
set([(12, 3, 27), (True, 60)])
````

#### `remove()`.
Busca y elimina al elemento que se ingresa como argumento. En caso de no encontrarlo, envía un mensaje de error.
```

```

- ```<set>``` es el objeto tipo _set_ del que se desea eliminar un elemento.
- ```<elemento a eliminar>``` es el elemento que se desea eliminar del conjunto

**Ejemplos:**

````python
conjunto = {1, 'dos', 3.0, '4'}
````

````python
conjunto
````

* La siguiente celda ejecutará el método ```remove()```, eliminando el elemento ```3.0``` del conjunto, ya que ```3``` y ```3.0``` son equivalentes.

* La siguiente celda intentará eliminar el elemento ```3``` del conjunto, el cual ya no existe, lo que desencadenará un error de tipo ```KeyError```.

### Métodos de evaluación de conjuntos.

**Nota:** Los diagramas de Venn ilustran los casos en los que la evaluación da por resultado _True_.

#### ```isdisjoint()```.
Evalúa si dos conjuntos no se intersectan. Si no existen elementos compartidos entre el objeto tipo _set_ al que pertenece el método y el objeto _set_ que se ingresa  como argumento, se regresa el valor _True_. De lo contrario se regresa _False_.

![img/isdisjoint.png](img/isdisjoint.png)

**Ejemplos:**

````python
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6}
conjunto_c = {5, 6, 7}
````

* La siguiente celda evaluará si ```conjunto_a``` y ```conjunto_b``` son disjuntos (no comparten elementos), lo que regresará ```False```.

````python
conjunto_a.isdisjoint(conjunto_b)
````

* La siguiente celda evaluará si ```conjunto_a``` y ```conjunto_c``` son disjuntos, lo que regresará ```True```.

````python
conjunto_a.isdisjoint(conjunto_c)
````

#### El método `issubset()`.
Si el objeto tipo `set` al que pertenece el método es subconjunto del objeto `set` que se ingresa como argumento, se regresa el valor `True`. De lo contrario se regresa `False`.

![img/issubset.png](img/issubset.png)

**Ejemplos:**


````python
conjunto_a = {1, 2, 3, 4}
conjunto_b = {1, 2, 4}
conjunto_c = {1}
````

* La siguiente celda evaluará si ```conjunto_a``` es subconjunto de ```conjunto_b```, lo que regresará ```False```.

````python
conjunto_a.issubset(conjunto_b)
````

* La siguiente celda evaluará si ```conjunto_b``` es subconjunto de ```conjunto_a```, lo que regresará ```True```.

````python
conjunto_b.issubset(conjunto_a)
````

* La siguiente celda evaluará si ```conjunto_c``` es subconjunto de ```conjunto_a```, lo que regresará ```True```.

````python
conjunto_c.issubset(conjunto_a)
````

## Objetos tipo _frozenset_.

Los objetos de tipo frozenset representan conjuntos inmutables.

### Definición de un objeto tipo _frozenset_.

Los objetos tipo _frozenset_ se definen mediante la función _frozenset()_ a la cual se le ingresa un objeto *iterable*.

``` 
frozenset(<objeto iterable>)
```
**Ejemplo:**


````python
frozenset([1, 'dos', 3.0, '4'])
````

* La siguiente celda creará un objeto ```frozenset``` a partir de un ```set``` con elementos numéricos equivalentes y una cadena.

````python
frozenset({2, 2., 1 + 1, "dos"})
````

* La siguiente celda creará un objeto ```frozenset``` a partir de un ```dict```, obteniendo un conjunto inmutable con las claves del diccionario.

````python
frozenset({'id_1': 3, 'id_2': 4, 'id_3': 5, 'id_4': 6})
````

### Métodos de ```frozenset```.

Debido a que los objetos ```frozenset``` son inmutables, estos no cuentan con métodos que modifiquen al conjunto.

* ```copy()```
* ```issubset()```
* ```issuperset()```
* ```union()```
* ```difference()```
* ```intersection()```
* ```symmetric_difference()```
