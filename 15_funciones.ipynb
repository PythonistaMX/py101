# Funciones.

Las funciones en *Python* son objetos que pueden ser ejecutados al ser invocados y que realizan ciertas operaciones que fueron asignadas al momento de ser definidas.

```
<func>(<arg 1>, <arg 2>, ..., <arg n>)
```

Donde:

* ```<func>``` es el nombre de una función.
* ```<arg 1>, <arg 2>, ..., <arg n>``` es una sucesión objetos que se ingresan como argumentos. 

## Características de las funciones.

* Las funciones pueden aceptar argumentos que serán asignados a los parámetros definidos en dichas funciones. Es necesario que para cada parámetro definido en una función se ingrese un argumento.
* Cada  función tiene su propio ámbito.
* Las funciones siempre regresan un objeto como resultado de su ejecución. Si no se indica lo contrario, las funciones regresan ```None```.

**Ejemplos:**

* La siguiente celda ejecutará la función ```dir()``` sin ingresar argumentos. El resultado será el contenido del espacio de nombres del intérprete.

````python
dir()
````

````python
len("Hola, mundo.")
````

* La siguiente celda ejecutará a la función ```print()```, ingresando al objeto ```"Hola, Mundo."``` como argumento. 

* La función ```print()``` despliega un mensaje ```Hola, Mundo.``` en la salida estándar del intérprete, pero en realidad regresa ```None```.

### Ámbito global.

El espacio de nombres del intérprete de *Python* corresponde al ámbito global.

#### La función ```globals()```.

La función ```globals()``` regresa el contenido del espacio de nombres del ámbito global como un objeto de tipo ```dict```.

Cuando se invoca la función ```dir()``` sin argumentos desde el intérprete, ésta regresa un objeto de tipo ```list``` con el listado de nombres del ámbito global.

### Ámbitos locales.

Cada función genera su propio espacio de nombres cada vez que es invocada. Cada uno de estos espacios de nombres es un ámbito local.

**Nota:** Los parámetros definidos en una función pertenecen al ámbito local.

#### La función ```locals()```.

La función ```locals()``` regresa el contenido del espacio de nombres del ámbito local como un objeto de tipo ```dict```. Cuando se invoca la función ```dir()``` sin argumentos desde una función, ésta regresa un objeto de tipo ```list``` con el listado de nombres del ámbito local.

**Ejemplo:**

* Se definirá la función ```ambitos()```, la cual desplegará:
   * El contenido de su ámbito local mediante ```locals()``` y ```dir()```.
   * El espacio de nombres del ámbito global con ```globals()```. 

````python
def ambitos():
    lista = [1, 2, 3]
    nulo = None
    print('Espacio de nombres en el ámbito local:')
    print('%s\n%s\n' %(locals(), dir()))
    print('----------------')
    print('Espacio de nombres en el ámbito global:')
    print(globals())
````

### Definiendo nombres en el ámbito global con la palabra reservada ```global```.

Es posible que una función pueda ligar un objeto al espacio de nombres del ámbito global mediante el uso de la expresión ```global``` con la siguiente sintaxis.

```
def <func>():
    global <nombre>
    <nombre> = <objeto>
```

Donde:
* `<func>` es el nombre de una función.

* ```<nombre>``` es el nombre que se quiere añadir al ámbito global.
* ```<objeto>``` es un objeto ligado a ```<nombre>```.

**Advertencia:** El uso de `global` se considera una mala práctica en la gran mayoría de los casos.

**Ejemplo:**

* La siguiente celda define en el ámbito global al nombre ```mi_nombre``` ligado al objeto ```'Juan'```.

````python
mi_nombre = "Juan"
````

* La función ```nombre_global()``` definirá de forma global al objeto ```"Hola"``` con nombre ```mi_nombre```. 

````python
def nombre_global():
    global mi_nombre
    mi_nombre = "Hola"
````

* Al ejecutar la función ```nombre_global()```, el nombre ```mi_nombre``` será sustituído por el objeto ```"Hola"```.

* La siguiente celda ejecutará la función ```nombre_global()```, la cual modificará el nombre ```mi_nombre``` en el ámbito global, ligándolo al objeto ```"Hola"```.

````python
nombre_global()
````

* La siguiente celda desplegará el objeto ligado al nombre ```mi_nombre```, el cual ahora es ```"Hola"``` tras haber ejecutado ```nombre_global()```.

````python
mi_nombre
````

### Argumentos posicionales.

Al llamar una función con cierto número de parámetros definidos, es necesario ingresar el mismo  número de valores. A estos valores se les denomina argumento.

Los argumentos posicionales permiten asignar valores a cada parámetro confome son ingresados.

```
<func>(<arg 1>, <arg 2>, ...  <arg n>)
```

Donde: 

* ```<func>``` es el nombre de una función existente.
* ```<arg 1>, <arg 2>, ...  <arg n>``` es una sucesión de objetos.

De este modo, el valor del primer argumento se asigna al primer parámetro, el segundo argumento al segundo parámetro y así sucesivamente.

### Parámetros con valor por defecto.

Es posible asignar valores por defecto a los parámetros de una función mediante el operador de asignación `=`.

* Llamada flexible: Si todos los parámetros tienen un valor por defecto, no es obligatorio pasar argumentos al invocar la función.

* Sustitución posicional: Los argumentos que se pasen al llamar a la función sustituirán a los valores por defecto en el orden en que aparezcan.

* Orden de definición: Los parámetros con valores por defecto siempre deben ir después de los parámetros que no los tienen.

```
def <func>(<param 1>=<valor 1>, <param 2>=<valor 2>, ...  <param n>=<valor n>):
   ...
   ...
```

Donde: 

* ```<func>``` es el nombre de la función a definir.
* ```<param 1>, <param 2>, ... <param 3>``` corresponden a nombres de parámetros.
* ```<valor 1>, <valor 2>, ... <valor 3>``` corresponden a valores que se le asignan a cada parámetro. 

**Ejemplos:**

La función ```suma()``` define los argumentos:
* ```primero=1```
* ```segundo=3```

````python
def suma(primero=1, segundo=3):
    '''Despliega la suma de dos objetos'''
    print(primero + segundo)
````

### Captura de múltiples argumentos con ```*args```.

Es posible definir un parámetro que acepte un número indeterminado de argumentos y que éstos queden guardados dentro de un objeto tipo ```tuple```. Para esto, basta preceder al nombre del parámetro con un solo asterisco ```*```.
```
def <func>(<param 1>, <param 2>, ... <param n>, *args) 
```

Donde:
* `<func>` es el nombre de una función.

* ```<param 1>, <param 2>, ... <param n>``` son parámetros a los que se les asigna sólo un argumento.
* `args` es un parámetro que es capaz de recibir cualquier número de argumentos, los cuales serán contenidos un objeto de tipo ```tuple```. El nombre `args` se utiliza por convención.

El parámetro `*args` debe colocarse al final.

**Ejemplo:**

* La siguiente celda definirá la función ```promedio()```, incluyendo al parámetro ```*args```.
   * Se desplegará el contenido del parámetro ```args```.
   * Se realizarán algunas operaciones con las funciones ```sum()``` y ```len()``` usando al parámetro ```args```.

````python
def promedio(*args):
    '''Función que despliega el promedio de un número 
    arbitrario de argumentos.'''
    print('Muestras:', args)
    '''Calcula el promedio de la muestra correspondiente a 
    todos los parámetros ingresados.'''
    promedio = sum(args) / len(args)
    print('El promedio de la muestra de %d elementos es %.3f.' %(len(args), promedio))
````

* La siguiente celda ejecutará a la función ```promedio()``` con 8 argumentos, los cuales serán asignados como elementos de un objeto de tipo ```tuple``` al parámetro ```args``` en el ámbito de la función ```promedio()``` .

````python
promedio(1, 3, 5, 8, 11, 24, 90, 29)
````

* La siguiente celda ejecutará a la función ```promedio()``` con 3 argumentos, los cuales serán asignados como elementos de un objeto de tipo ```tuple``` al parámetro ```args``` en el ámbito de la función ```promedio()``` .

````python
promedio(14, 38, 1)
````

**Ejemplo:**

* La siguiente celda definirá la función ```promedio_titulo()```, incluyendo los parámetros ```titulo``` y ```args```.

````python
def promedio_titulo(titulo, *args):
    '''Función que despliega un titulo que se ingresa como primer argumento
    y despliega el promedio del resto de argumentos ingresados.'''
    promedio = sum(args)/len(args)
    print('Título:', titulo)
    print('El promedio de la muestra de %d elementos es %.3f.' %(len(args), promedio))
````

### *Keyword Arguments*.

Además de los argumentos posicionales, es posible indicarle a una función el parámetro específico al que se le asigna un valor mediante los *keyword arguments*.

```
<param> = <valor>
```
Donde: 

* `<param>` es el nombre de un parámetro definido en una función.
* `<valor>`es el valor que se le asignará `<param>`.


```
<func>(<arg 1>, <arg 2>, ... <arg n>, <kwarg 1>, <kwarg 2>, ..., <kwarg n>)
```

Donde:
* `<func>` es una función con múltiples parámetros definidos.
* `<arg 1>, <arg 2>, ... <arg n>` son argumentos posicionales.
* `<kwarg 1>, <kwarg 2>, ..., <kwarg n>` son *keyword arguments*.

**Notas:** 
* Los *keyword arguments* siempre deben escribirse después de todos los argumentos posicionales.
* Si se asigna un argumento posicional a un parámetro y se intentara asignar un valor mediante un *keyword argument* al mismo parámetro, se desencadenará un error de tipo `TypeError`.

**Ejemplo:**

* Se utilizará la función `suma_tres()`.

````python
def suma_tres(primero, segundo=2, tercero=5):
    '''Función que define tres parámetros de los cuales
      sólo dos tienen valores por defecto.'''
    print(primero + segundo + tercero)
````

### Captura de múltiples *keyword arguments* con ```**kwargs```.

Es posible definir un parámetro que acepte un número indeterminado de *keyword arguments* adicionales. Estos serán capturados en un objeto de tipo `dict`. Para lograr esto, se debe preceder el nombre del parámetro con un doble asterisco `**`.

```
def <func>(<params>, **kwargs) 
```

Donde:

* ```<params>``` son parámetros definidos de forma explícita.
* ```kwargs``` es el nombre del parámetro que contendrá el diccionario de argumentos extra. El nombre *kwargs* se utiliza por convención.

**Nota:** Los *keyword arguments* contenidos en `kwargs` son únicamente aquellos que no corresponden a los parámetros predefinidos (`<params>`).

**Ejemplo:**

* La función ```superficie()``` define al parámetro ```**kwargs```.
     * Se calculará ```superficie``` dependiendo del valor del elemento ```kwargs["tipo"]```.
     * Se hace referencia a:
         * ```kwargs["base"]``` y ```kwargs["altura"]``` en caso de que ```kwargs["tipo"]``` sea ```'Rectángulo'``` o ```'Triángulo'```.
         * ```kwargs["radio"]``` en caso de que ```kwargs["tipo"]``` sea ```'Círculo'```.
     * En caso de que ```kwargs["tipo"]``` sea distinto a ```'Rectángulo'```, ```'Triángulo'``` o ```'Círculo'```, ```superficie``` será igual a ```'indefinido'```.

````python
def superficie(**kwargs):
    '''Calcula la superficie de una figura geométrica si los parámetros  ingresados
       coinciden.'''
    print("Argumentos: ", kwargs)
    if kwargs["tipo"] == "Rectángulo":
        superficie = float(kwargs["base"]) * float(kwargs["altura"])
    elif kwargs["tipo"] == "Triángulo":
        superficie = float(kwargs["base"]) * float(kwargs["altura"]) / 2
    elif kwargs["tipo"] == "Círculo":
        superficie = float(kwargs["radio"]) ** 2 * 3.14259265
    else:
        print("No puedo calcular la superficie.")
        superficie = "indefinido"
    print("La superficie del %s es %s." % (kwargs["tipo"].lower(), superficie))
````

* La siguiente celda ejecutará la función ```superficie()``` ingresando los argumentos ```base=22```, ```altura=30``` y ```tipo="Rectángulo"```.
* La función ```superficie()``` asignará al parámetro ```kwargs``` el objeto ```{'base': 22, 'altura': 30, 'tipo': 'Rectángulo'}```.

````python
superficie(base=22, altura=30, tipo="Rectángulo")
````

* La siguiente celda ejecutará la función ```superficie()``` ingresando los argumentos ```tipo="Círculo"``` y ```radio=35```.
* La función ```superficie()``` asignará al parámetro ```kwargs``` el objeto ```{'tipo': 'Círculo', 'radio': 35}```.

````python
superficie(tipo="Círculo", radio = 35)
````

* La siguiente celda ejecutará la función ```superficie()``` ingresando los argumentos ```base=22```, ```altura=30``` y ```tipo="Rombo"```.
* La función ```superficie()``` asignará al parámetro ```kwargs``` el objeto ```{'base': 22, 'altura': 30, 'tipo': 'Rombo'}```.

````python
superficie(base=22, altura=30, tipo="Rombo")
````

* La siguiente celda ejecutará la función ```superficie()``` ingresando los argumentos ```base=22```, ```altura=30```, ```tipo="Rectángulo"``` y ```radio=6```.
* La función ```superficie()``` asignará al parámetro ```kwargs``` el objeto ```{'base': 22, 'altura': 30, 'tipo': 'Rectángulo', 'radio': 6}```.

````python
superficie(base=22, altura=30, tipo="Rectángulo", radio=6)
````

* La siguiente celda ejecutará la función ```superficie()``` ingresando los argumentos ```base=22```, ```altura=30``` y ```tipo="Círculo"```.
* La función desencadenará un error de tipo ```KeyError``` debido a que espera que el parámetro ```kwargs``` contenga un elemento con clave ```radio```.

````python
superficie(base=22, altura=30, tipo="Círculo")
````

### La palabra reservada  ```return```.

La expresión ```return``` se utiliza para regresar uno o varios objetos específicos a su ámbito superior y acto seguido dar por terminada la ejecución de la función de forma similar a ```break```. 

Pueden incluirse varios ```return``` en una función, pero sólo se ejecutará la primera que se encuentre. La sintaxis es la siguiente:

```
return <objeto 1>, <objeto 2>, ..., <objeto n> 
```

Donde:

* ```<objeto i>``` puede ser cualquier objeto de Python.

En el caso de que ```return``` regrese una secuencia de objetos, estos estarán contenidos en un objeto de tipo ```tuple```.

**Ejemplo:**

* La siguiente celda define a la función ```regresa_promedio()```, la cual regresará el resultado de varias expresiones que usan el parámetro ```args```.

````python
def regresa_promedio(*args):
    return len(args), sum(args) / len(args)
````

* La siguiente celda ejecutará a la función ```regresa_promedio()``` ingresando ```1, 3, 5, 8, 11, 24, 90, 29``` como argumentos. La función regresará ```(8, 21.375)```.

````python
regresa_promedio(1, 3, 5, 8, 11, 24, 90, 29)
````

* La siguiente celda ejecutará a la función ```regresa_promedio()``` ingresando ```1, 3, 5, 8, 11, 24, 90, 29``` como argumentos. El objeto que regrese la función será asignado al nombre ```media```.

````python
media = regresa_promedio(1, 3, 5, 8, 11, 24, 90, 29)
````

* El nombre ```media``` corresponde al objeto ```(8, 21.375)```.

````python
media
````

* La siguiente celda desplegará un mensaje formateado con el número de elementos y el promedio calculado, utilizando los valores de la tupla ```media```.

````python
print('El promedio de la muestra de %d elementos es %.3f.' %(media))
````

### Cerraduras.

Una cerradura o *closure* es una función definida dentro de otra función que conserva el acceso a los nombres del ámbito envolvente incluso después de que la función externa ha terminado de ejecutarse.

Una cerradura no es lo mismo que el valor regresado por cualquier función. La cerradura ocurre cuando una función interna hace referencia a nombres del ámbito local de la función externa y dicha función interna es devuelta o conservada para usarse después.

**Ejemplo:**


* La siguiente celda define a la función ```crear_saludo()```, la cual contiene a la función interna ```saludar()```.
* La función interna ```saludar()``` hace referencia al nombre ```mensaje```, definido en el ámbito local de ```crear_saludo()```.
* Al ejecutar ```crear_saludo()```, ésta regresará a la función ```saludar()``` conservando el acceso al nombre ```mensaje```.

````python
def crear_saludo():
    mensaje = 'Hola'

    def saludar(nombre):
        return f'{mensaje}, {nombre}.'

    return saludar
````

* La siguiente celda asignará al nombre ```saludo_hola``` la función que regresa ```crear_saludo()```.
* Aun cuando ```crear_saludo()``` ya terminó su ejecución, la función ```saludo_hola``` conservará acceso al nombre ```mensaje```.

````python
saludo_hola = crear_saludo()
````

* La siguiente celda ejecutará a ```saludo_hola()``` con el argumento ```'Mundo'```.
* El resultado mostrará que la función interna conserva el valor de ```mensaje``` aun después de que terminó la ejecución de ```crear_saludo()```.

````python
saludo_hola('Mundo')
````
