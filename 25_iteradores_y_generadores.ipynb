# Iteradores y generadores.

Uno de los conceptos más poderosos en *Python* es el uso de iteradores y generadores.

## Iteradores.

Los "iteradores" y "generadores" son objetos que cuentan con el método ```__next__()```, el cual regresa una serie de objetos de uno en uno cada vez que es invocado.

La implementación del método ```__next__()``` corresponde a la función ```next()```.

## Iterables.

Un objeto "*iterable*" es un objeto que cuenta con al menos una implementación del método ```__iter__()```, el cual da por resultado un objeto "iterador" cuando es invocado.

La implementación del método ```__iter__()``` corresponde a la función ```iter()```.

**Ejemplo:**

Por ejemplo, el tipo ```list``` es *iterable* de la siguiente forma.

````python
lista = ['1', 2, 'tres', 4.0]
````

Los objetos de tipo ```list``` cuentan con un método ```__iter__()```, pero no con un método ```__next__()```.

````python
hasattr(lista, "__iter__")
````

* La siguiente celda confirmará que ```lista``` no implementa el método ```__next__()```.

````python
hasattr(lista, "__next__")
````

Sin embargo, el objeto creado a partir del método `__iter__()` sí cuenta con un método `__next__()` y por lo tanto se trata de un iterador.

````python
iterador = iter(lista)
print(iterador)
````

* La siguiente celda ejecutará la función ```type()``` sobre ```iterador``` para mostrar el tipo del objeto creado.

````python
type(iterador)
````

* La siguiente celda verificará que ```iterador``` sí implementa el método ```__next__()```.

````python
hasattr(iterador, "__next__")
````

El objeto ```iterador``` va a regresar cada elemento del objeto ```lista``` de uno en uno cada vez que se invoque la función ```next()``` y cuando ya haya dado todos, regresará un mensaje de error de tipo ```StopIteration``` indicando que el iterador está vacío.

````python
next(iterador)
````

* La siguiente celda regresará el primer elemento producido por ```iterador```.

````python
next(iterador)
````

* La siguiente celda regresará el segundo elemento producido por ```iterador```.

````python
next(iterador)
````

* La siguiente celda regresará el tercer elemento producido por ```iterador```.

````python
next(iterador)
````

* La siguiente celda regresará el cuarto elemento producido por ```iterador```.

````python
next(iterador)
````

* La siguiente celda intentará obtener un quinto elemento. Como el iterador ya estará vacío, se desencadenará ```StopIteration```.

## Implementación de las iteraciones con ```for``` ... ```in```.

La estructura ```for``` ... ```in``` es la implementación de *Python* para aprovechar a los objetos *iterables*.

Cuando se utiliza esta estructura, el intérprete de *Python* realiza los siguientes pasos:

1. Busca un método ```__iter__()``` que se ajuste al requerimiento y crea un objeto iterador.
2. Realiza una sucesión de ejecuciones de ```__next__()``` para el objeto iterador resultante hasta que éste quede vacío.

Por lo que el siguiente código:

* La siguiente celda recorrerá una lista mediante la estructura ```for``` ... ```in``` y desplegará cada elemento.

````python
for i in ['1', 2, 'tres', 4.0]:
    print(i)
````

Sería similar al siguiente:

* La siguiente celda ilustra una implementación equivalente usando explícitamente ```__iter__()``` y ```__next__()```.

````python
iterador = ['1', 2, 'tres', 4.0].__iter__()
while True:
    try: 
        i = iterador.__next__()
        print(i)
    except StopIteration:
        break
````

## Generadores.

Los generadores son objetos iteradores. Es decir, producen valores uno a uno y cuentan con el método ```__next__()```.

En *Python*, estos objetos pueden obtenerse principalmente de dos maneras:

* Mediante funciones generadoras, que utilizan la palabra reservada ```yield```.
* Mediante expresiones generadoras, como las estudiadas posteriormente en el tema de comprensiones.

Una función generadora sí es invocable (*callable*), pero el objeto generador que produce es un iterador, no una función.

La palabra reservada ```yield``` regresa un objeto desde el ámbito local de una función al ámbito superior de ésta, pero a diferencia de ```return``` la función no termina en ese punto, sino que queda suspendida para reanudar su ejecución en la siguiente llamada a ```next()``` sobre el generador.

**Ejemplo:**

Para ilustrar el uso de ```yield``` se creará una función generadora que produzca números primos a partir del número ```2``` de forma ascendente, uno a la vez.

* La siguiente celda definirá la función generadora ```gen_primos()```.

* Las celdas posteriores crearán un objeto generador, comprobarán su tipo, verificarán que implementa ```__next__()``` y consumirán varios de sus valores.

````python
def gen_primos():
    """ Generador de números primos."""
    
    contador = 1
    lista_primos = []
               
    # Comienza una secuencia infinita.
    while True:
        es_primo = True
        contador += 1
        if len(lista_primos) > 0:
            for primo in lista_primos: 
                if contador % primo == 0:
                    es_primo = False
                    break
        if es_primo:
            lista_primos.append(contador)
            yield contador
````

* La siguiente celda creará un objeto generador a partir de la función ```gen_primos()```.

````python
generador = gen_primos()
````

````python
type(generador)
````

````python
hasattr(generador, '__next__')
````

* La siguiente celda regresará el tercer número primo producido por ```generador```.

````python
next(generador)
````

* La siguiente celda continuará consumiendo el generador y regresará otro valor primo.

````python
next(generador)
````

* La siguiente celda continuará consumiendo el generador y regresará otro valor primo.

* La siguiente celda continuará consumiendo el generador y regresará otro valor primo.

* La siguiente celda seguirá consumiendo el generador dentro de un ciclo ```for``` para desplegar los valores restantes.

````python
def gen_finito(limite=100):
    contador = 1
    lista_primos=[]
               
    # Comienza una secuencia infinita.
    while contador < limite:
        es_primo = True
        contador += 1
        if len(lista_primos) > 0:
            for primo in lista_primos: 
                if contador % primo == 0:
                    es_primo = False
                    break
        if es_primo:
            lista_primos.append(contador)
            yield contador
````

* La siguiente celda definirá una función generadora finita llamada ```gen_finito()```, que producirá números primos hasta un límite dado.

````python
finito = gen_finito(70)
````

* La siguiente celda creará el generador ```finito``` usando el límite ```70```.

````python
next(finito)
````

* La siguiente celda regresará el primer valor producido por ```finito```.

````python
type(finito)
````

* La siguiente celda mostrará el tipo del objeto ```finito```.

````python
for item in finito:
    print(item)
````

* La siguiente celda recorrerá ```finito``` con un ciclo ```for``` y desplegará los valores restantes hasta agotar el generador.

````python
next(finito)
````

* La siguiente celda intentará obtener otro valor de ```finito``` después de haberlo agotado, por lo que se desencadenará ```StopIteration```.

````python
for item in gen_finito(1000):
    print(item)
````

* La siguiente celda recorrerá directamente un nuevo generador creado por ```gen_finito(1000)```.

En este caso, al invocar directamente ```gen_finito(1000)``` dentro del ciclo ```for```, se crea un nuevo generador para esa iteración completa. Esta idea es consistente con las expresiones generadoras estudiadas después: en ambos casos los valores se producen bajo demanda, en lugar de materializar toda la secuencia desde el inicio.
