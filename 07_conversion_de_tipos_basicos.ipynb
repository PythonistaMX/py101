## ObtenciĂłn del tipo de dato de un objeto.

La funciĂłn ```type()``` regresa el tipo de dato o la clase a la que pertenece un objeto el cual es ingresado como argumento con la siguiente sintaxis:

```
type(<objeto>)
```

Donde: 

* ```<objeto>``` es cualquier objeto.

**Ejemplos:**

* Las siguientes celdas utilizarĂˇn la funciĂłn ```type()``` para conocer el tipo de datos del que se trata.

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```"Hola"```, el cual corresponde a ```str```.

````python
type("Hola")
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```b"Hola"```, el cual corresponde a ```bytes```.

````python
type(b"Hola")
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```12```, el cual corresponde a ```int```.

````python
type(12)
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```12.```, el cual corresponde a ```float```.

````python
type(12.)
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```23j```, el cual corresponde a ```complex```.

````python
type(23j)
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```True```, el cual corresponde a ```bool```.

````python
type(True)
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```[1, 2, 3]```, el cual corresponde a ```list```.

````python
type([1, 2, 3])
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```(1, 2, 3)```, el cual corresponde a ```tuple```.

````python
type((1, 2, 3))
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```{1, 2, 3}```, el cual corresponde a ```set```.

````python
type({1, 2, 3})
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```{'uno': '1'}```, el cual corresponde a ```dict```.

````python
type({'uno': '1'})
````

* La siguiente celda desplegarĂˇ el tipo al que pertenece el objeto ```None```, el cual corresponde a ```NoneType```.

````python
type(None)
````

## La funciĂłn ```int()```.

Esta funciĂłn transforma un objeto compatible que es ingresado como argumento a un objeto tipo ```int```. 
 
La sintaxis es la siguiente:

```
int(<objeto>)
```

Donde: 

* ```<objeto>``` es el objeto que serĂˇ convertido a un objeto de tipo ```int```.

### Particularidades.

* Es posible convertir objetos de tipo ```str``` que representen correctamente a un nĂşmero entero.
* Los objetos de tipo ```float``` son truncados en la parte entera. 
* ```True``` es convertido en ```1```.
* ```False``` es convertido en ```0```. 
* La funciĂłn ```int()``` no es compatible con objetos tipo ```complex```, lo que originarĂˇ un error de tipo ```TypeError```.
* La funciĂłn ```int()``` no es compatible con ```None```, lo que originarĂˇ un error de tipo ```TypeError```.

**Ejemplos:**

* La siguiente celda convertirĂˇ en un objeto ```int``` al objeto de tipo ```bool``` que se ingresa como argumento.

````python
int(True)
````

* El argumento de la siguiente celda contiene una representaciĂłn correcta de un entero negativo, por lo que la funciĂłn ```int()``` podrĂˇ realizar la conversiĂłn correctamente.

````python
int("-12")
````

* El argumento de la siguiente celda contiene una representaciĂłn correcta de un nĂşmero real, por lo que la funciĂłn ```int()``` no podrĂˇ realizar la conversiĂłn correctamente y regresarĂˇ un error de tipo ```ValueError```.

````python
int("28.8")
````

````python
int(b"12")
````

* El argumento de la siguiente celda no contiene una representaciĂłn correcta de un nĂşmero entero, por lo que la funciĂłn ```int()``` serĂˇ incapaz de realizar la conversiĂłn y desencadenarĂˇ un error de tipo ```ValueError```.

````python
int('Hola')
````

* Los argumentos de las siguiente celdas contienen objetos tipo ```float```, por lo que la funciĂłn ```int()``` truncarĂˇ el valor a enteros.

````python
int(5.6)
````

````python
int(-5.3)
````

* Los argumentos de las siguientes celdas son objetos de tipo ```complex```, por lo que la funciĂłn ```int()``` serĂˇ incapaz de realizar la conversiĂłn y desencadenarĂˇ un error de tipo ```TypeError```.

````python
int(12 + 45.2j)
````

````python
int(-5j)
````

* El argumento de la siguiente celda es ```None```, el cual no puede ser representado por algĂşn valor numĂ©rico, lo que desencadenarĂˇ un error de tipo ```TypeError```.

````python
int(None)
````

## La funciĂłn ```complex()```.

Transforma a un objeto compatible a uno de tipo ```complex```  y puede ser usada con las siguientes sintaxis:

### Ingresando pares numĂ©ricos como argumentos.

```
complex(<real>, <imaginario>)
```
Donde:

* ```<real>``` corresponde al primer argumento que se ingresa a la funciĂłn ```complex()``` y puede ser un objeto de tipo ```int```, ```float``` e incluso ```bool```. Este serĂˇ usado como el componente real del nĂşmero complejo.
* ```<imaginario>``` corresponde al segundo argumento que se ingresa a la funciĂłn ```complex()``` y puede ser un objeto de tipo ```int```, ```float``` e incluso ```bool```. Este serĂˇ usado como el componente imaginario del nĂşmero complejo. Su valor por defecto es ```0```.


### Ingresando una cadena de caracteres como argumento.

```
complex(<cadena>)
```
Donde:

* ```<cadena>``` corresponde a un objeto de tipo ```str``` que contenga una representaciĂłn correcta de un nĂşmero ```complex```.

**Ejemplos:**

* Las siguientes celdas definen argumentos numĂ©ricos para la funciĂłn ```complex()```.

````python
complex(3.5, 2)
````

````python
complex(8.3)
````

````python
complex(False)
````

````python
complex(True)
````

````python
complex(True, True)
````

* Las siguientes celdas definen argumentos de tipo ```str``` cuyos contenidos son compatibles con la funciĂłn ```complex()```.

````python
complex("23+5j")
````

````python
complex("23")
````

* El argumento de la siguiente celda es la representaciĂłn de una expresiĂłn que al ser evaluada da por resultado un objeto de tipo ```complex```. Sin embargo, la funciĂłn ```complex()``` no tiene la capacidad de evaluar expresiones, originĂˇndose un error de tipo ```ValueError```.

````python
complex("23 + 5j")
````

## La funciĂłn ```str()```.

La funciĂłn ```str()``` permite realizar transformaciones a objetos tipos ```str``` mediante las siguientes sintaxis:

### TransformaciĂłn de objetos compatibles.

```
str(<objeto>)
```
Donde:

* ```<objeto>``` es un objeto compatible con la funciĂłn ```str()```; prĂˇcticamente todos los tipos bĂˇsicos son compatibles con esta funciĂłn y el resultado es una cadena de caracteres que representa al objeto en cuestiĂłn.


### TransformaciĂłn de un objeto de tipo ```bytes``` o  ```bytearray``` a ```str```.

```
str(<objeto>, encoding=<codificaciĂłn>)
```
Donde:

* ```<objeto>``` es un objeto  de tipo ```bytes``` o ```bytearray``` cuyo contenido serĂˇ convertido a una cadena de caracteres.
* ```<codificaciĂłn>``` corresponde al [tipo de codificaciĂłn](https://docs.python.org/3/library/codecs.html#standard-encodings) a la que se convertirĂˇ la cadena de bytes. Por lo general es ```"utf-8"``` o ```"ascii" ```. 

En caso de que no se defina el atributo ```encoding```, el resultado serĂˇ una representaciĂłn del objeto ```bytes```.

**Ejemplos:**

* Las siguientes celdas regresarĂˇn un objeto ```str``` con la representaciĂłn de los objetos que se ingresan como argumentos.

````python
str(True)
````

````python
str(False)
````

````python
str(12 + 3.5j)
````

````python
str({'nombre': 'Juan'})
````

````python
str(None)
````

````python
str(b'Saludos')
````

* La siguiente celda convertirĂˇ el contenido del objeto ```b'Saludos'``` en una cadena de caracteres usando la codificaciĂłn *ASCII*.

````python
str(b'Saludos', encoding='ascii')
````

* La siguiente celda convertirĂˇ el contenido del objeto ```bytearray(b'Saludos')``` en una cadena de caracteres usando la codificaciĂłn *ASCII*.

````python
str(bytearray(b'Saludos'), encoding='ascii')
````

* La siguiente celda convertirĂˇ el contenido del objeto ```b'G\xc3\xb6del'``` en una cadena de caracteres usando la codificaciĂłn *UTF-8*.

````python
str(b'G\xc3\xb6del', encoding='utf-8')
````

* La siguiente celda intentarĂˇ convertir el contenido del objeto ```b'G\xc3\xb6del'``` en una cadena de caracteres usando la codificaciĂłn *ASCII*. Sin embargo, debido a que dicha codificaciĂłn no es compatible con el contenido, se desencadenarĂˇ un error de tipo ```UnicodeDecodeError```. 

````python
str(b'G\xc3\xb6del', encoding='ascii')
````

## La funciĂłn ```bytes()```.

Transforma a un objeto que es ingresado como argumento en un objeto de tipo ```bytes```.

Las sintaxis son las siguientes:

```
bytes(<cadena>, encoding=<codificaciĂłn>)
```
Donde:

* ```<cadena>``` es un objeto  de tipo ```str``` cuyo contenido serĂˇ convertido a una cadena de bytes.
* ```<codificaciĂłn>``` corresponde al tipo de codificaciĂłn a la que se convertirĂˇ la cadena de caracteres. Por lo general es ```"utf-8"``` o ```"ascii" ```. En caso de que no se defina este argumento, se desencadenarĂˇ un error de tipo ```UnicodeError```.


**Ejemplos:**

La siguiente celda regresarĂˇ la representaciĂłn en bytes del objeto ```'Saludos'``` usando la codificaciĂłn *ASCII*.

````python
bytes('Saludos', encoding='ascii')
````

````python
bytes('GĂ¶del', encoding='utf-8')
````

````python
bytes('GĂ¶del', encoding='latin-1')
````

* La siguiente celda intentarĂˇ realizar la representaciĂłn en bytes del objeto ```'GĂ¶del'``` usando la codificaciĂłn *ASCII*. Sin embargo, dicha codificaciĂłn no contiene un cĂłdigo para el caracter ```Ă¶```, por lo que se generarĂˇ un error de tipo ```UnicodeEncodeError```.
