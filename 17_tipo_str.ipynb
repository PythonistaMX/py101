# El Tipo ```str```.

Este tipo de datos es una colección ordenada, indexable numéricamente e inmutable de caracteres Unicode.

Para definir un objeto de tipo ```str``` es necesario delimitarlo ya sea con comillas ```"``` o con apóstrofes ```'```.

Además de los caracteres Unicode, es posible incluir secuencias de escape antecedidas por la diagonal invertida ```\``` tales como las que se muestran en la siguiente tabla.

|Secuencia|Despliegue|
|:-------:|:--------:|
|```\n```    |Retorno de línea|
|```\t```    |Tabulador |
|```\"```    |Comillas  |
|```\'```    |Apóstrofe |
|```\\```   |Diagonal invertida|
|```\xNN```  |Carácter cuyo código hexadecimal corresponde a *NN*|
|```\uNNNN```  |Carácter Unicode cuyo punto de código hexadecimal corresponde a *NNNN*|

Estas secuencias de escape serán interpretadas por funciones de despliegue tales como ```print()```. 

**Ejemplos:**

* Las siguientes celdas definen objetos ```str```.

* La siguiente celda define un objeto ```str``` con caracteres del alfabeto chino.

* La siguiente celda define un objeto ```str``` que incluye una comilla simple en su contenido, utilizando comillas dobles como delimitadores.

````python
"Vamos a MacDonald's"
````

Los caracteres contenidos en un objeto de tipo ```str``` no pueden ser modificados usando el operador ```=``` ni eliminados con ```del```.

### El método  ```partition()```.

Este método busca dentro del objeto ```str``` el texto contenido en otro objeto ```str``` que se ingresa como argumento, realizando la búsqueda de izquerda a derecha. 

* En caso de que encuentre una coincidencia, regresará un objeto de tipo ```tuple``` el cual contiene:
   * Un objeto ```str``` que contiene el texto a la derecha del texto encontrado.
   * Un objeto ```str``` que contiene al texto encontrado.
   * Un objeto ```str``` que contiene el texto a la izquerda del texto encontrado.

* En caso de que no exista una coincidencia, regresará un objeto de tipo ```tuple```el cual contiene:
   * Al objeto ```str``` original.
   * Dos objetos ```str``` vacíos.
   
```
<objeto str>.partition(<cadena>)
```

Donde:

* ```<cadena>``` es un objeto ```str``` que contiene al texto que será buscado.

**Ejemplos:**

* Las siguientes celdas ejemplifican distintos resultados del método ```partition()```.

````python
"Hugo, Paco, Luis".partition(", ")
````

* La siguiente celda aplicará el método ```partition()``` buscando el separador ```'+'```, el cual no existe en el objeto, por lo que regresará el texto original flanqueado por dos cadenas vacías.

````python
"Hugo, Paco, Luis".partition("+")
````

### Métodos de mapeo de caracteres.

* ```<mapa 2>``` es un objeto ```str``` en el que cada carácter está ligado al carácter de ```<mapa 1>``` con el mismo índice.

**Ejemplo:**

###  El método ```encode()```.

Este método permite transformar el contenido de un objeto ```str``` de una codificación a un objeto de tipo ```bytes```. 

```
str.encode(<argumentos>)
```

El formato de codificación se define mediante el parámetro ```encoding```. Por lo general se utiliza el argumento ```encoding="utf-8"```.

Es común que no todos los caracteres tengan una correspondencia de una codificación a otra, por lo que se pueden generar errores de conversión. Se pueden definir cuatro formas de gestionar estos errores mediante el parámetro ```errors```, al cual se le pueden asignar los siguientes argumentos. 

* ```errors='strict'```, el cual emite un error de codificacion. Si no se indica, esta es la forma por defecto de gestionar los errores de codificación.
* ```errors='ignore'```, el cual ignora el carácter en cuestión y prosigue con el siguiente.
* ```errors='replace'``` el cual sustituye al carácter en cuestión por un signo de interrogación ```?```.
* ```errors='xmlcharrefreplace'``` el cual sustituye el carácter por el valor correspondiente a su [valor de entidad de carácter de XML](https://es.wikipedia.org/wiki/Anexo:Referencias_a_entidades_de_caracteres_XML_y_HTML). 

**Ejemplos:**

* Las siguientes celdas ejemplifican distintos resultados del método ```encode()```.

````python
'Holåé'.encode(encoding="ascii", errors='ignore')
````

* La siguiente celda codificará ```'Holåé'``` en ```ASCII``` con la política ```'replace'```, sustituyendo los caracteres no compatibles con el símbolo ```?```.

````python
'Holåé'.encode(encoding="ascii", errors='replace')
````

* La siguiente celda codificará ```'Holåé'``` en ```ASCII``` con la política ```'xmlcharrefreplace'```, sustituyendo los caracteres no compatibles con referencias de entidades XML.

````python
'Holåé'.encode(encoding="ascii", errors='xmlcharrefreplace')
````

* La siguiente celda intentará codificar ```'Holåé'``` en ```ASCII``` con la política ```'strict'```, lo que desencadenará un error de tipo ```UnicodeEncodeError``` al encontrar caracteres no compatibles.

````python
'Holåé'.encode(encoding="ascii", errors='strict')
````

* La siguiente celda codificará ```'Holåé'``` usando la codificación ```Latin-1```, la cual soporta los caracteres ```å``` y ```é```.

````python
'Holåé'.encode(encoding="Latin-1", errors='strict')
````

* La siguiente celda codificará ```'Holåé'``` usando la codificación ```utf-8```, la cual soporta una amplia variedad de caracteres Unicode.

````python
'Holåé'.encode(encoding="utf-8", errors='strict')
````

### El método ```format()```.

Este método regresa un objeto ```str``` al que se le pueden aplicar  múltuiples procedimientos de formateo y sustitución del texto. 

La sustitución se realiza mediante el usos de llaves ```{ }``` dentro del objeto ```str``` original. Los elementos a sustituir se ingresan como argumentos separados por comas.

Este método permite formatear y alinear los elementos a los que sustituye de muchas maneras.

En el sitio [https://pyformat.info/](https://pyformat.info/) es posible explorar las distintas formas en las que se puede utilizar dicho método.

#### Control de la sustitución con el método ```format()```.

Si no se especifica, cada elemento será sustituido de forma sucesiva de izquierda a derecha. Pero también es posible indicar cual elemento sustituir por mediante un índice numérico o mediante una clave, de tal forma que los argumentos ingresados son indexados  como objetos de tipo ```tuple``` o de tipo ```dict```.
