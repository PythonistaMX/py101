La forma más básica de distribuir código en Python es mediante la copia de los archivos fuente. Sin embargo, Python ofrece mecanismos en su biblioteca estándar para empaquetar aplicaciones completas en un solo archivo ejecutable.

En este capítulo exploraremos cómo estructurar un paquete para que sea ejecutable y cómo utilizar el módulo ```zipapp``` para crear archivos distribuibles, sin depender de herramientas externas.


## Estructura para ejecución.

Para que un paquete sea ejecutable directamente por Python (por ejemplo ```python -m mi_paquete```), debe incluir un archivo especial llamado ```__main__.py``` en la raíz del paquete.

Este archivo será el punto de entrada que se ejecutará cuando el paquete sea invocado como un script.

La estructura del directorio [```paquete```](paquete/) en este entorno es la siguiente:

```
paquete/
    __init__.py
    __main__.py
    promedios.py
    primos/
        __init__.py
        funciones.py
```


## Archivo ```__main__.py```.

El contenido del archivo [```paquete/__main__.py```](paquete/__main__.py) es el siguiente. Nótese cómo importa funcionalidad relativa al paquete.


!cat paquete/__main__.py


## Ejecución del paquete.

Podemos ejecutar el paquete directamente usando la opción ```-m``` de Python.


!python3 -m paquete 10 20 30


## Creación de un archivo pyz con ```zipapp```.

El módulo ```zipapp``` (introducido en Python 3.5) permite comprimir todo el paquete en un solo archivo con extensión ```.pyz``` que puede ser ejecutado directamente por el intérprete.

La sintaxis básica es:
```bash
python -m zipapp <directorio_del_paquete>
```


````python
!python3 -m zipapp paquete
````

````python
Esto generará un archivo ```paquete.pyz```.
````

!ls -F *.pyz


Ahora podemos ejecutar este archivo directamente:


!python3 paquete.pyz 5 7 9
