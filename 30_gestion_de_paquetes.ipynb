## La biblioteca de paquetes de *Python*.

Existe una enorme cantidad de bibliotecas y aplicaciones escritas en *Python* y publicadas en el índice oficial de paquetes [https://pypi.org/](https://pypi.org/).

Ese índice se conoce como *Python Package Index* o *PyPI* y es la fuente habitual desde la que se descargan paquetes de terceros.

## El gestor de paquetes ```pip```.

```pip``` es la herramienta estándar para instalar y administrar paquetes desde PyPI. Sigue siendo la base de muchas otras herramientas del ecosistema.

### Disponibilidad de ```pip```.

En Python moderno, ```pip``` suele venir incluido con la instalación oficial de *Python*.

Si fuera necesario habilitarlo o repararlo, normalmente se utiliza ```ensurepip``` o la reinstalación del intérprete. En sistemas administrados por una distribución de *GNU/Linux* también puede instalarse mediante el gestor de paquetes del sistema.

En entornos actuales se recomienda usar ```python -m pip``` en lugar de invocar ```pip``` directamente, para dejar explícita la versión de Python con la que se está trabajando.

**Ejemplo:**

A continuación se muestra una forma habitual de verificar que ```pip``` está disponible para *Python* 3.

````python
!python3 -m pip --version
````

## Operaciones básicas con ```pip```.

Para gestionar paquetes con ```pip``` es habitual usar la línea de comandos.

Cuando se modifica la instalación de *Python* o un entorno virtual, es necesario contar con los permisos adecuados dentro de ese entorno.

```
python -m pip <subcomando> <argumentos>
```

### Instalación de paquetes.

Para instalar uno o varios paquetes con ```pip``` se utiliza el subcomando ```install``` seguido del nombre de uno o varios paquetes.

```bash
$ python -m pip install <paquete 1> <paquete 2> ... <paquete n>
```

Una de las ventajas de ```pip``` es que calcula e instala también las dependencias requeridas por los paquetes solicitados.

## Control de dependencias.

Instalar paquetes no basta para mantener un proyecto estable. También es importante controlar qué versiones se usan y poder reproducir el mismo entorno en otra máquina o en otro momento.

### Especificación de versiones.

*Python* permite expresar restricciones de versión al instalar dependencias.

```bash
$ python -m pip install "requests>=2.31,<3"
```

Esto reduce el riesgo de que una actualización incompatible cambie el comportamiento del proyecto.

### Archivos ```requirements.txt``` y ```constraints.txt```.

Un archivo ```requirements.txt``` enumera las dependencias que deben instalarse en un entorno.

```bash
$ python -m pip install -r requirements.txt
```

Cuando se necesita mayor control, también pueden utilizarse archivos ```constraints.txt``` para limitar versiones sin declarar directamente qué paquetes deben instalarse.

### Archivos de bloqueo.

En proyectos modernos es frecuente usar archivos de bloqueo que registran versiones exactas de las dependencias y de sus dependencias indirectas.

Ese enfoque mejora la reproducibilidad del entorno. Herramientas como ```uv``` o ```Poetry``` pueden generar y mantener esos archivos automáticamente a partir de la definición del proyecto.

**Ejemplo:**

La siguiente celda actualizará ```pip``` usando la forma recomendada dentro de un entorno interactivo o *notebook*.

````python
%pip install --upgrade pip
````

### Actualización de paquetes.

En caso de que se requiera actualizar uno o varios paquetes, se utiliza el subcomando ```install``` con el parámetro adicional ```--upgrade```.

```bash
$ python -m pip install --upgrade <paquete 1> <paquete 2> ... <paquete n>
```

Al actualizar dependencias conviene revisar las versiones resultantes, especialmente en proyectos compartidos, para evitar diferencias entre entornos.

### Listado de paquetes instalados.

El gestor ```pip``` permite desplegar un listado de los paquetes instalados mediante el subcomando ```list```.

```bash
$ python -m pip list
```

Este tipo de consulta es útil para auditar el entorno y verificar qué versiones están realmente instaladas.

**Ejemplo:**

````python
!pip list
````

### Desinstalación de paquetes.

Para desinstalar un paquete se utiliza el subcomando ```uninstall```.

```bash
$ python -m pip uninstall <paquete 1> <paquete 2> ... <paquete n>
```

**Nota:** Antes de desinstalar paquetes, ```pip``` pide confirmación. Para realizar la operación sin preguntar, se utiliza la opción ```-y```.

En la práctica, estas operaciones conviene realizarlas dentro de un entorno virtual para no alterar la instalación principal de *Python* del sistema.

### Documentación adicional sobre ```pip```.

El gestor ```pip``` cuenta con más subcomando y opciones avanzadas que pueden ser consultadas en https://pip.pypa.io/en/stable/.

## Herramientas actuales para gestionar paquetes.

Aunque ```pip``` sigue siendo fundamental, hoy existen herramientas de más alto nivel o más especializadas que resuelven tareas frecuentes con mayor comodidad.

### ```uv```.

[```uv```](https://docs.astral.sh/uv/) es una herramienta moderna y muy rápida para crear entornos, resolver dependencias, instalar paquetes y ejecutar comandos relacionados con proyectos Python.

Puede usarse como reemplazo práctico de varios flujos basados en ```pip``` y ```venv```, y trabaja bien con proyectos modernos basados en ```pyproject.toml```.

Una de sus ventajas más importantes es que facilita mantener dependencias sincronizadas y reproducibles mediante archivos de bloqueo del proyecto.

**Ejemplos:**

```bash
$ uv add requests
$ uv lock
$ uv sync
$ uv run python *script*.py
```

### *pipx*.

[```pipx```](https://pipx.pypa.io/stable/) está pensado para instalar aplicaciones de línea de comandos escritas en *Python* en entornos aislados, sin mezclar sus dependencias con las del proyecto actual.

Es especialmente útil para utilidades como formateadores, "linters" o generadores de proyectos.

**Ejemplo:**

```bash
$ pipx install black
```

### Otras herramientas de proyecto.

También existen herramientas como [*Poetry*](https://python-poetry.org/) o [*Hatch*](https://hatch.pypa.io/latest/) , que ayudan a gestionar dependencias, metadatos del proyecto, construcción de paquetes y publicación. En este curso basta con conocer que forman parte del ecosistema moderno, aunque no se estudien en detalle.
