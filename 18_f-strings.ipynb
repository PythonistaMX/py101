# *f-strings*.

Como se ha visto en capítulos anteriores, es posible formatear cadenas de caracteres utilizando la sintaxis del carácter de escape ```%``` y el método ```str.format()```.



Ambas técnicas sustituyen dentro de un texto el resultado de las expresiones que se ingresan ya sea como una tupla en el caso del carácter de escape ```%``` o como argumentos del método ```str.format()```.

**Ejemplos:**

* La siguiente celda definirá las variables `radio` y  `pi` con valores numéricos.

````python
radio = 12
pi = 3.14
````

* La siguiente celda desplegará un mensaje de texto que incluye el resultado de dos expresiones utilizando el especificador de formato ```%f``` en ambos casos, lo que añadirá al texto una representación de punto flotante del resultado de cada expresión.

````python
print('La superficie de una circunferencia de radio %f es de %f' % (radio, pi * radio ** 2))
````

* La siguiente celda regresará una cadena de caracteres que incluye:

  *  El resultado de la expresión `radio` utilizando el especificador de formato ```%d``` (despliega como entero).

  * El resultado de la expresión `pi * radio ** 2` utilizando el especificador de formato ```%f``` (despliega como punto flotante).


````python
'La superficie de una circunferencia de radio %d es de %f' %(radio, pi * radio ** 2)
````

* La siguiente celda desplegará un mensaje de texto cuyo contenido será construido usando el método `format()` incluyendo las expresiones `radio` y `pi * radio ** 2` como argumentos. En este caso, el resultado de cada expresión se desplegará utilizando la representación por defecto de cada tipo de dato.

````python
print('La superficie de una circunferencia de radio {} es de {}'.format(radio, pi * radio ** 2))
````

## Las *f-strings*.

La [*PEP-498*](https://peps.python.org/pep-0498/) define el uso de cadenas de caracteres especiales, conocidas como *f-strings*, las cuales permiten sustituir expresiones directamente en un texto. La representación de cada expresión debe estar delimitada por llaves ```{``` ```}```.





```

f'...{<exp 1>}...{<exp 2>}...{<exp n>}...'

```



```

f"...{<exp 1>}...{<exp 2>}...{<exp n>}..."

```



Donde:



* ```<exp x>``` es la representación de una expresión de *Python*. 

**Ejemplos:**

* La siguiente celda define las variables `nombre` e `email` con objetos de tipo `str`.

````python
nombre = "Juan"
email = "juan@ejemplo.com"
````

* La siguiente celda regresa una cadena de caracteres que incluye el texto definido en `nombre`.

````python
f'Hola. Mi nombre es {nombre}.'
````

* La siguiente celda desplegará el texto definido en `email` con un formato específico.

````python
print(f'Mi dirección de correo electrónico es {email}.')
````

* La siguiente celda definirá las variables numéricas `radio` y `pi`.

````python
radio = 12
pi = 3.14
````

* La siguiente celda desplegará un mensaje de texto que incluye el resultado de las expresiones `radio` y `radio**2 * pi` con un formato específico.

````python
print(f'La superficie de una circunferencia de radio {radio} es de {pi * radio ** 2}')
````

## Formato de valores en *f-strings*.



Es posible aplicar formatos a los valores interpolados dentro de las *f-strings* utilizando una sintaxis similar a la de ```format()```, añadiendo dos puntos ```:``` seguido del especificador de formato.



La documentación de los formatos aceptados en las *f-strings* se encuentra en el siguiente enlace:



* https://docs.python.org/3/library/string.html#formatspec


**Ejemplo:**

* La siguiente celda desplegará un mensaje de texto que incluye:

  * El resultado de la expresión `radio` representado con el formato por defecto.

  * El resultado de la expresión `pi * radio ** 2` representado con un formato de punto flotante con 2 decimales.

````python
print(f'La superficie de una circunferencia de radio {radio} es de {pi * radio ** 2:.2f}')
````

## Las *f-strings* a partir de *Python 3.12*.



A partir de *Python 3.12*, la sintaxis de las *f-strings* se ha relajado, permitiendo:

* Reutilizar las mismas comillas dentro de la expresión.

* Usar caracteres de escape (como `\n` o `\"`) dentro de la expresión.

* Escribir expresiones de múltiples líneas y comentarios.

**Ejemplos:**

* La siguiente celda definirá el objeto `persona` de tipo `dict`.

````python
persona = {"nombre": "Juan", "edad": 30}
````

* En la siguiente celda se utilizarán comillas para delimitar la *f-string* y comillas dentro de la *f-string*.

````python
print(f"Nombre: {persona['nombre']}")  # Válido en todas las versiones (comillas distintas)
````

* En la siguiente celda se utilizarán comillas tanto para delimitar la *f-string* como para las claves del diccionario. En versiones previas a *Python 3.12*, esta sintaxis desencadenaría un error. 

**Nota:** Este ejemplo requiere *Python 3.12* o superior para ejecutarse correctamente.

````python
print(f"Nombre: {persona["nombre"]}")  # Válido en Python 3.12+ (mismas comillas)
````

* La siguiente celda ejemplifica una sintaxis que no era permitida antes de *Python 3.12*:

  * El uso de caracteres de escape dentro de una expresión en una *f-string*.

  * El uso de comillas tanto delimitadoras de la *f-string* como dentro de la expresión.

**Nota:** Este ejemplo requiere *Python 3.12* o superior para ejecutarse correctamente.

````python
frutas = ["manzana", "banana", "cereza"]
print(f"Salto: {"\n".join(frutas)}") # carácter de escape en la expresión
````

## *f-strings* de varias líneas.



Al combinar las *f-strings* con triples comillas ```f"""..."""``` es posible definir plantillas de texto multilínea con interpolación de expresiones. Esto resulta útil para generar mensajes, reportes o cualquier texto estructurado cuyo contenido dependa de variables.

**Ejemplo:**

* La siguiente celda define las variables `nombre`, `saldo` y `fecha`.

````python
nombre = "Ana"
saldo = 1523.75
fecha = "10 de abril de 2026"
````

* La siguiente celda define la variable `estado_cuenta` como una *f-string* de varias líneas que incluye las variables definidas en la celda anterior, formateando el valor de `saldo` como un número con dos decimales y utilizando el formato de fecha para `fecha`.

````python
estado_cuenta = f"""
Estimada {nombre}:

A continuación se muestra su estado de cuenta al {fecha}.

  Saldo disponible: ${saldo:,.2f}

Gracias por su preferencia.
"""
````

* La siguiente celda desplegará el mensaje generado por la *f-string* de varias líneas definida en `estado_cuenta`.

````python
print(estado_cuenta)
````
