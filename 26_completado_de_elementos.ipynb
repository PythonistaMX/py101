# Comprensiones y expresiones generadoras

*Python* permite construir nuevos objetos a partir de un *iterable*, una expresión y, opcionalmente, una condición. Esta técnica suele conocerse por su nombre en inglés *comprehension*.

Este tema se relaciona con lo estudiado anteriormente sobre funciones, expresiones ```lambda``` y programación funcional: en todos esos casos se describe cómo transformar datos a partir de reglas y expresiones, en lugar de detallar paso a paso cada operación intermedia.

Las comprensiones nativas se aplican directamente a objetos de tipo:

* ```list```.
* ```dict```.
* ```set```.

Además, *Python* permite usar expresiones generadoras, las cuales producen valores bajo demanda. A partir de ellas es posible construir otros objetos, como ```tuple```, ```frozenset```, ```bytes``` o ```bytearray```, usando sus funciones constructoras.

## Comprensiones en objetos de tipo _list_.

La sintaxis es la siguiente para un objeto de tipo _list_.
```
[<expresión> for <nombre> in <iterable> if <expresión lógica>]
```
El proceso de completado es el siguiente:
* Realiza la iteración definida por la expresión _for_ ... *in*.
* A cada elemento iterado le aplica la condición lógica, en caso de existir.
* Si la condición lógica se cumple, añade el resultado de la expresión aplicada al elemento.

Es válido no incluir una expresión condicional.

**Ejemplos:**

* La siguiente celda muestra que, en este caso concreto, el resultado es equivalente a ejecutar la función ```list()``` sobre la misma cadena.

* La siguiente celda construirá una lista con los múltiplos de ```5``` correspondientes a los números pares del ```1``` al ```20```.

* La siguiente celda generará una lista con las consonantes de ```'Parangaricutirimicuaro'``` convertidas a mayúsculas.

````python
[x for x in 'Hola']
````

* La siguiente celda muestra que, en este caso concreto, el resultado es equivalente a ejecutar la función ```list()``` sobre la misma cadena.

````python
list('Hola')
````

````python
[letra.upper() for letra in 'Parangaricutirimicuaro' if letra.lower() not in ['a', 'e', 'i', 'o', 'u']]
````

Usando el último ejemplo, el código sin utilizar una comprensión de listas sería similar a lo siguiente:

### Completado con expresiones ternarias.

Es posible utilizar expresiones ternarias para el completado de elementos con la siguiente sintaxis:

```
[<expresión ternaria> for <nombre> in <iterable>]
```

**Ejemplos:**

* La siguiente celda conservará los números pares y multiplicará por ```5``` los impares.

* La siguiente celda convertirá a mayúsculas las consonantes de ```'chapultepec'``` y dejará en minúsculas las vocales.

````python
[numero if numero % 2 == 0 else 5 * numero for numero in range(1, 21)]
````

````python
lista = []
for numero in range(1,21):
    lista.append(numero if numero % 2 == 0 else 5 * numero)
lista
````

* La siguiente celda convertirá a mayúsculas las consonantes de ```'chapultepec'``` y dejará en minúsculas las vocales.

````python
[letra.upper() if letra.lower() not in ['a', 'e', 'i', 'o', 'u'] else letra.lower() for letra in 'chapultepec']
````

## Expresiones generadoras y conversión a ```tuple```.

Cuando se escribe una expresión entre paréntesis con la forma ```(<expresión> for <nombre> in <iterable> if <condición>)```, el resultado no es una ```tuple```, sino un generador.

Un generador produce sus elementos bajo demanda. Si se desea obtener una ```tuple``` con esos valores, es necesario consumir el generador y convertirlo mediante la función ```tuple()```.

**Ejemplo:**

* La siguiente celda definirá un generador que produce los múltiplos de ```5``` correspondientes a los números pares del ```1``` al ```20```.

* La siguiente celda confirmará que el objeto creado es de tipo generador.

* La siguiente celda desplegará la representación del generador, no sus elementos.

* La siguiente celda iterará sobre el generador e imprimirá cada elemento producido.

* La siguiente celda volverá a crear el generador, ya que el anterior habrá sido consumido durante la iteración.

* La última celda convertirá el generador en una ```tuple```.

````python
generador = (5 * x for x in range(1, 21) if x % 2 == 0)
````

* La siguiente celda confirmará que el objeto creado es de tipo generador.

````python
type(generador)
````

* La siguiente celda desplegará la representación del generador, no sus elementos.

````python
print(generador)
````

* La siguiente celda iterará sobre el generador e imprimirá cada elemento producido.

* La siguiente celda volverá a crear el generador, ya que el anterior habrá sido consumido durante la iteración.

````python
generador = (5 * x for x in range(1, 21) if x % 2 == 0)
````

* La siguiente celda convertirá el generador en una ```tuple```.

````python
tuple(generador)
````

## Comprensiones en objetos de tipo _dict_.

En una comprensión de diccionario, cada iteración produce un par ```clave: valor```.

Sintaxis:
```
{<clave>: <valor> for <nombre> in <iterable> if <expresión lógica>}
```

Es válido omitir la condición lógica cuando no sea necesaria.

**Ejemplo:** 

* La siguiente celda definirá en ```campos``` los nombres que se utilizarán como claves del diccionario resultante.

* La siguiente celda ejecutará una comprensión de diccionario que solicitará un valor con ```input()``` para cada clave. Esta celda es interactiva y depende de que el *notebook* esté en ejecución.

* La siguiente celda repetirá la comprensión, pero guardará el resultado en la variable ```registro```.

* La última celda desplegará el contenido de ```registro```.

````python
campos = ('nombre', 'primer apellido', 'segundo apellido', 'correo')
````

* La siguiente celda ejecutará una comprensión de diccionario que solicitará un valor con ```input()``` para cada clave. Esta celda es interactiva y depende de que el *notebook* esté en ejecución.

````python
{campo: input(f'Ingrese {campo}: ') for campo in campos}
````

````python
persona = {}
for campo in campos:
    persona[campo] = input(f'Ingrese {campo}: ')
persona
````
