## Funciones anidadas.

*Python* permite definir funciones dentro de otras funciones.

Una cerradura o *closure* aparece cuando una función interna conserva el acceso a los nombres del ámbito envolvente incluso después de que la función externa ha terminado de ejecutarse.

No toda función anidada es, por sí sola, una cerradura; ésta ocurre cuando la función interna es devuelta o conservada para usarse después.

**Ejemplo:**

* La función ```lista_primos()``` contiene a su vez a la función ```es_primo()```.
* La función ```es_primo()``` solamente existe dentro del ámbito local de la función  ```lista_primos()```.

````python
def lista_primos(limite=100):
    '''Genera una lista de los números primos comprendidos
    entre el 2 y el valor de limite.'''
    
    #La lista inicia con el número 2.
    lista = [2]
   
    def es_primo(numero):
        '''Valida si numero es divisible entre un numero 
        contenido en lista.'''
        
        # Iterará cada número primo contenido en lista.
        for primo in lista:
        
            # Si numero es divisible entre primo, regresa False.
            if numero % primo == 0:
                return False
            
        # Si numero no es divisible entre ningún número
        # contenido en lista, regresa True.
        return True
    
    # Iterará todos los número enteros entre 3 y limite.
    for numero in range(3, limite + 1):
        
        # Si es_primo(numero) regresa True,  añade 
        # el valor de numero a lista.
        if es_primo(numero):
            lista.append(numero)
    # Regresa la lista de numeros primos.        
    return lista
````

* La siguiente celda ejecutará la función ```lista_primos()``` ingresando el argumento ```1050```.
* El resultado será una lista con los números primos comprendidos entre ```2``` y ```1050```.

````python
lista_primos(1050)
````

* La siguiente celda intentará ejecutar a la función ```es_primo()``` desde el intérprete.
* Debido a que ```es_primo()``` sólo existe dentro del ámbito local de ```lista_primos()```, se desencadenará un error de tipo ```NameError```.

````python
es_primo()
````

* En el ejemplo anterior se definió a la función ```es_primo()``` dentro de la función ```lista_primos()```. 
* Como se puede observar, el nombre ```lista``` está en el espacio de nombres de ```lista_primos()```, pero al estar en un entorno superior al ámbito de ```es_primo()```, ésta función puede acceder a ```lista```.

## Recursividad.

*Python* permite hacer llamadas recursivas a una función. Es decir, que la función se invoque a sí misma.

Cada vez que una función se invoca recursivamente, *Python* crea un nuevo marco de ejecución con un ámbito local propio para esa llamada. La función sigue siendo la misma, pero cada invocación conserva sus propios valores locales hasta regresar un resultado.

**Ejemplo:**

```
1! = 1
2! = 2 * 1! = 2
3! = 3 * 2! = 6
4! = 4 * 3! = 24
5! = 5 * 4! = 120
```

````python
def factorial(n):
    if n == 1:
        return 1
    else:
        return  n * factorial(n - 1)
````

* La siguiente celda ejecutará la función ```factorial()``` ingresando el argumento ```5```.
* El resultado será ```120```.

````python
factorial(5)
````

En este caso, la función ```factorial()``` se invoca recursivamente y, en cada llamada, el valor del parámetro ```n``` decrece en ```1``` hasta alcanzar el valor de ```1```, momento en el que se regresa dicho valor.

A partir de ese caso base, cada llamada pendiente multiplica su valor de ```n``` por el resultado regresado por la llamada recursiva de nivel inferior hasta llegar nuevamente a la invocación inicial.

Ahora se incluirán algunas modificaciones al ejemplo anterior para ilustrar el proceso.

````python
def factorial(n):
    print(f'En este ámbito, el valor de n es {n}.')
    if n == 1:
        print('Llegó a 1.')
        print('Regresa 1! = 1.')
        return 1
    else:
        fact = n * factorial(n - 1)
        print(f'Regresa {n} * {n - 1}! = {fact}.')
        return fact
````

* La siguiente celda ejecutará nuevamente a la función ```factorial()``` con el argumento ```5```.
* En este caso se desplegarán mensajes intermedios para ilustrar el proceso de la recursión.

````python
factorial(5)
````

## Funciones de orden superior.

Las funciones de orden superior son funciones que aceptan funciones como argumentos y a su vez regresan funciones.

**Ejemplo:**

* La función ```p()``` transforma al texto ingresado como argumento, correspondiente al parámetro ```texto```, en el código *HTML* de un elemento ```<p>```.

````python
def p(texto):
    '''Crea el código HTML de un bloque <p>.'''
    return f'<p>{texto}</p>'
````

* La siguiente celda ejecutará la función ```p()``` ingresando el argumento ```'Hola, Mundo.'```.
* El resultado será una cadena con el código *HTML* correspondiente a un elemento ```<p>```.

````python
p('Hola, Mundo.')
````

* La siguiente celda desplegará la documentación del objeto ```p```.

````python
help(p)
````

* La función ```h1()``` transforma al texto ingresado como argumento, correspondiente al parámetro ```texto```, en el código *HTML* de un elemento ```<h1>```.

````python
def h1(texto):
    '''Genera el código HTML de un bloque <h1> '''
    return "<h1>" + texto + "</h1>"
````

* La siguiente celda ejecutará la función ```h1()``` ingresando el argumento ```'Título'```.
* El resultado será una cadena con el código *HTML* correspondiente a un elemento ```<h1>```.

````python
h1('Título')
````

* La siguiente celda desplegará la documentación del objeto ```h1```.

````python
help(h1)
````

* La función ```doc()``` regresa una cadena de caracteres que representa a un documento *HTML5*, incluyendo el texto ingresado como argumento como el contenido de ```<body>```.

````python
def doc(texto):
    '''Permite crear un documento HTML5.'''
    plantilla = '<!DOCTYPE html>\n\
    <html>\n\
      <head>\n\
        <title>Página</title>\n\
      </head>\n\
      <body>\n\
        {}\n\
      </body>\n\
    </html>'
    return plantilla.format(texto)
````

* La siguiente celda ejecutará la función ```doc()``` ingresando el argumento ```'<p>Hola</p>'```.
* El resultado será una cadena que representa a un documento *HTML5* completo.

````python
doc('<p>Hola</p>')
````

* La siguiente celda imprimirá en pantalla el resultado de ```doc('<p>Hola</p>')```.

````python
print(doc('<p>Hola</p>'))
````

* La siguiente celda imprimirá el documento *HTML5* generado por ```doc('<p>Hola</p>')```.

````python
print(doc(p('Hola')))
````

* La siguiente celda imprimirá el resultado de aplicar ```doc()``` al valor que regresa ```p('Hola')```.

````python
print(doc(h1('Hola')))
````

* La siguiente celda imprimirá el resultado de aplicar ```doc()``` al valor que regresa ```h1('Hola')```.

* La función ```html()``` es una función de orden superior que:
    * Recibe una función como argumento para el parámetro ```func```. 
    * Define a la función anidada ```doc()```, la cual crea una cadena de caracteres a partir de una plantilla y del resultado de invocar ```func()``` pasándole el argumento ```texto```. 

````python
def html(func):
    '''Regresa una función que construye un documento HTML5 
    a partir del texto que regresa el parámetro func.'''
    def doc(texto):
        '''Permite crear un documento HTML5.'''
        
        plantilla = '<!DOCTYPE html>\n\
        <html>\n\
          <head>\n\
            <title>Página</title>\n\
          </head>\n\
          <body>\n\
            {}\n\
          </body>\n\
        </html>'
        
        # Se regresará la plantilla con lo que regrese
        # la función` correspondiente al parámetro func.
        return plantilla.format(func(texto))
    
    # Regresa a la función doc
    return doc
````

* La siguiente celda desplegará la documentación del objeto ```html```.

````python
help(html)
````

* La siguiente celda ejecutará a ```html(p)```.
* El resultado será una función que espera un texto y regresa un documento *HTML5*.

````python
html(p)
````

* La siguiente celda ejecutará a la función que regresa ```html(p)``` ingresando el argumento ```"Hola"```.

````python
html(p)("Hola")
````

* La siguiente celda imprimirá el documento *HTML5* que se obtiene al ejecutar ```html(p)("Hola")```.

````python
print(html(p)("Hola"))
````

* La siguiente celda imprimirá el documento *HTML5* que se obtiene al ejecutar ```html(h1)("Hola")```.

````python
print(html(h1)("Hola"))
````

## Decoradores.

Los decoradores son un recurso de *Python* que permiten aplicar una función de orden superior a otra función con la siguiente sintaxis.

```
@<nombre de función de orden superior>
def <nombre>(<params>):
    ...
    ...
```
Donde:
* El símbolo ```@``` indica que se va a aplicar un decorador a la función que se define a continuación.
* ```<nombre de función de orden superior>``` es el nombre de la función de orden superior que se va a aplicar a la función que se define a continuación.
* ```<nombre>``` es el nombre de la función a la que se le aplicará el decorador.
* ```<params>``` son los parámetros de la función a la que se le aplicará el decorador

**Ejemplo:**

* Se utilizará el decorador de la función ```html()``` aplicado a la función ```p()```.

````python
@html
def p(texto):
    '''Crea el código HTML de un bloque <p>.'''
    return f'<p>{texto}</p>'
````

* La siguiente celda desplegará el objeto ligado al nombre ```p``` después de aplicar el decorador ```@html```.

````python
p
````

* La siguiente celda imprimirá el resultado de ejecutar ```p("Hola, Mundo.")```.

````python
print(p("Hola, Mundo."))
````

* La siguiente celda desplegará la documentación del objeto ```p``` después de haber sido decorado.

````python
help(p)
````

* La siguiente celda define a la función ```h1()``` usando el decorador ```@html```.

````python
@html
def h1(texto):
    '''Genera el código HTML de un bloque <h1> '''
    return "<h1>" + texto + "</h1>"
````

* La siguiente celda imprimirá el resultado de ejecutar ```h1('Título')```.

````python
print(h1('Título'))
````

* La siguiente celda desplegará la documentación del objeto ```h1``` después de haber sido decorado.

````python
help(h1)
````


## Definición de funciones con la declaración ```lambda```.

*Python* permite definir funciones en una sola línea mediante el uso de la expresión lambda con la siguiente sintaxis:

```
lambda <parámetro 1>,  <parámetro 2>...  <parámetro n> : <expresión>
```
Donde:
* El símbolo ```lambda``` indica que se va a definir una función anónima.
* ```<parámetro 1>,  <parámetro 2>...  <parámetro n>``` son los parámetros de la función anónima.
* El símbolo ```:``` separa la lista de parámetros de la expresión que se va a evaluar y regresar como resultado de la función anónima.
* ```<expresión>``` es la expresión que se va a evaluar y regresar como resultado de la función anónima.

A este tipo de funciones se les conoce como funciones lambda o funciones anónimas debido a que no requieren de un nombre para ser definidas.

Para nombrar a estas funciones se utiliza el operador de asignación ```=```.

**Ejemplos:**

````python
saluda = lambda t='extraño', w=50: f'Hola, {t}.'.center(w)
````

* La siguiente celda ejecutará la función ```type()``` ingresando el nombre ```saluda``` como argumento.
* El resultado será ```function```.

````python
type(saluda)
````

* La siguiente celda ejecutará a la función ```saluda()``` sin argumentos, por lo que se usarán los valores por defecto.

````python
saluda()
````

* La siguiente celda ejecutará a la función ```saluda()``` ingresando los argumentos ```'Mundo'``` y ```20```.

* La función ```es_par()``` regresa ```True``` si el argumento ingresado es un número par y regresa ```False``` si el argumento ingresado es un número impar.

````python
es_par = lambda n: True if n % 2  == 0 else False
````

* La siguiente celda ejecutará a la función ```es_par()``` ingresando el argumento ```2```.
* El resultado será ```True```.

````python
es_par(2)
````

* La siguiente celda ejecutará a la función ```es_par()``` ingresando el argumento ```3```.

* La función ```factorial()``` calcula el factorial de un número mediante recursividad.

````python
factorial = lambda n: n * factorial(n - 1) if n > 1 else 1
````

* La siguiente celda ejecutará a la función ```factorial()``` ingresando el argumento ```5```.
* El resultado será ```120```.

````python
factorial(5)
````
