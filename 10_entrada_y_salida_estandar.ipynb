## Salida estĂ¡ndar con la funciĂ³n ```print()```.

En Python 3, la funciĂ³n ```print()``` se utiliza para desplegar informaciĂ³n en la salida estĂ¡ndar.

La sintaxis es la siguiente:

```
print(<expresiĂ³n 1>, <expresiĂ³n 2>, ...<expresiĂ³n n>, sep=<separador>, end=<finalizador>)
```

Donde:
* ```<expresiĂ³n 1>```, ```<expresiĂ³n 2>```, ..., ```<expresiĂ³n n>``` son las expresiones que se desean desplegar.
* ```<separador>``` es un argumento opcional que especifica el separador entre las expresiones. Por defecto es un espacio.
* ```<finalizador>``` es un argumento opcional que especifica el finalizador que se agrega al final de la salida. Por defecto es un salto de lĂ­nea.

* La funciĂ³n ```print()``` evalĂºa y despliega una o varias expresiones.
* Si el resultado de la expresiĂ³n es un objeto ```str```, este es desplegado sin los apĂ³strofes o las comillas que lo delimitan.

**Ejemplos:**

* La siguiente celda define el nombre ```a``` con valor igual a ```2```.

````python
a = 2
````

* La siguiente celda evalĂºa la expresiĂ³n ```a```, por lo que desplegarĂ¡ ```2```.

````python
print(a)
````

* La siguiente celda desplegarĂ¡ el mensaje dentro del objeto ```"Hola"```.

````python
print("Hola")
````

* La salida serĂ¡ ```Hola Mundo```.

* En la siguiente celda la funciĂ³n ```print()``` desplegarĂ¡ el resultado de una expresiĂ³n de concatenaciĂ³n entre dos objetos de tipo ```str```. El resultado es un objeto ```str```.

## Despliegue con formato de los objetos ```str```.

Los objetos ```str```  permiten insertar las representaciones de otros objetos dentro de ellos mismos de 3 formas:

* El uso del caracter de escape ```%```.
* El mĂ©todo ```format()```, el cual se estudiarĂ¡ mĂ¡s adelante.
* A partir de Python 3.6 se incluye el usos de "f-strings", las cuales son descritas en el [*PEP 498*](https://www.python.org/dev/peps/pep-0498/) y se explorarĂ¡n mĂ¡s adelante.

### InserciĂ³n de objetos con formato en objetos  ```str``` mediante el caracter de escape ```%```.


Para intercalar valores dentro de un formato especĂ­fico de texto se utiliza el caracter sobre-escritura definido como el signo de porcentaje ```%``` seguido de algĂºn caracter que definirĂ¡ el modo de desplegar la expresiĂ³n correspondiente.

```
"...%<caracter>..." % expresiĂ³n 1) 
```

```
"...%<caracter 1>...%<caracter n>..." %(<expresiĂ³n 1>,...<expresiĂ³n n>)
```
|Caracter de escape|Modo de despliegue|
|:----------------:|:----------------:|
|```%s```|	cadena de texto|
|```%d```|	 entero|
|```%o```|	 octal|
|```%x```|	 hexadecimal|
|```%f```|	 punto flotante|
|```%e```|	 punto flotante en formato exponencial|    

El uso de ```%s```, equivale a aplicar la funciĂ³n ```str()``` al valor a desplegar.

Para desplegar el signo de porcentaje ```%``` se utiliza ```%%```.

**Ejemplos:**

* La siguiente celda definirĂ¡n los objetos ```pi``` y ```radio```.

````python
pi = 3.141592
radio = 2
````

* Las siguientes celdas desplegarĂ¡n en formatos distintos las expresiones correspondientes.

````python
print("El perĂ­metro de un cĂ­rculo de radio %d es %f." % (radio,
                                                         2 * radio * pi))
````

````python
print("El perĂ­metro de un cĂ­rculo de radio %d es %d." % (radio, 2 * radio * pi))
````

````python
print("El perĂ­metro de un circulo de radio %s es %s." % (radio, 2 * radio * pi))
````

````python
print("El valor de pi es %f." % (pi))
````

````python
print("El valor de pi es %e." % pi)
````

La siguiente celda incluye varias expresiones que definen a los objetos ```valor```, ```porciento``` y ```porcentaje```. Dichos objetos serĂ¡n desplegados dentro de un objeto ```str``` usando el caracter de escape ```%``` y el signo de porcentaje usando ```%%```.

````python
# Esta celda desplegarĂ¡ varias expresiones usando diversos formatos.
valor = 13
porciento = 15
porcentaje = (valor * porciento) / 100
print("El %d%% de %f es %f." % (porciento, valor, porcentaje))
````

#### Despliegue de cifras significativas.

Para desplegar un nĂºmero especĂ­fico de cifras significativas de un valor de punto flotante, se aĂ±ade un punto ```.``` y el nĂºmero de cifras a desplegarse despuĂ©s del signo de porcentaje ```%``` y antes del carĂ¡cter ```f``` o ```e```.

```
"<texto> %.<n>f <texto>"
```

Donde:

```<n>``` es el nĂºmero de decimales que serĂ¡n desplegados.

**Ejemplos:**

* La siguiente celda definirĂ¡n los objetos ```pi``` y ```radio```.

````python
pi = 3.14169265
radio = 2
````

* Las siguientes celdas desplegarĂ¡n de formas distintas las expresiones ```radio``` y ```2 * pi * radio```.

````python
print("El perĂ­metro de un cĂ­rculo de radio igual a %d es %f." % (radio, 2 * pi * radio))
````

````python
print("El perĂ­metro de un cĂ­rculo de radio igual a %d es %.2f." % (radio, 2 * pi * radio))
````

### Caracteres de escape.

Existen algunos caracteres que por su funciĂ³n o por la sintaxis de Python, tales como los apĂ³strofes, las comillas, los retornos de lĂ­nea; que deben de utilizar un "caracter de escape", para que puedan ser desplegados. Los caracteres de escape pueden ser introducidos despuĂ©s de una diagonal invertida ```\```. 

Estos caracteres de escape son usados en conjunto con la funciĂ³n ```print()```.

|Secuencia|Despliegue|
|:-------:|:--------:|
|```\n```    |Retorno de lĂ­nea|
|```\t```    |Tabulador |
|```\"```    |Comillas  |
|```\'```    |ApĂ³strofe |
|```\\```   |Diagonal invertida|
|```\xNN```  |Caracter que corresponde al nĂºmero hexadecimal *NN* en ASCII|
|```\uNNNN```  |Caracter que corresponde al nĂºmero hexadecimal *NN* en Unicode|

**Ejemplos:**

* Las siguientes celdas desplegarĂ¡n cadenas de caracteres que incluyen caracteres de escape.

````python
print("Primera lĂ­nea.\nSegunda lĂ­nea\t con tabulador.")
````

````python
print("Este es el signo de \"gato\" \x23.")
````

````python
print("Beta: \u00DF")
````

````python
print('I \u2764 YOU!')
````
