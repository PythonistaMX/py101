# Entrada y salida estándar.

En la actualidad existen muchas fuentes desde las que se puede obtener y desplegar la información que un sistema de cómputo consume, gestiona y genera. Sin embargo, para el intérprete de *Python* la salida por defecto (salida estándar) de datos es la terminal de texto y la entrada estándar es el teclado.

## La función ```print()```.

En *Python 3*, la función ```print()``` es capaz de enviar contenido de texto a la salida estándar.

La sintaxis es la siguiente:

```
print(<expresión 1>, <expresión 2>, ...<expresión n>, sep=<separador>, end=<finalizador>)
```

Donde:
* ```<expresión 1>```, ```<expresión 2>```, ..., ```<expresión n>``` son las expresiones que se desean desplegar.
* ```<separador>``` es un argumento opcional que especifica el separador entre las expresiones. Por defecto es un espacio.
* ```<finalizador>``` es un argumento opcional que especifica el finalizador que se agrega al final de la salida. Por defecto es un salto de línea.

* La función ```print()``` evalúa y despliega una o varias expresiones.
* Si el resultado de la expresión es un objeto ```str```, éste es desplegado sin los apóstrofes o las comillas que lo delimitan.

**Ejemplos:**

* La siguiente celda define el nombre ```a``` con valor igual a ```2```.

````python
a = 2
````

* La siguiente celda ejecuta `print(a)`: se desplegará `2` en la salida estándar y la variable `valor` recibirá el valor que regresa `print()`, que es `None`.

````python
valor = print(a)
````

* La siguiente celda desplegará el mensaje dentro del objeto ```"Hola"```.

````python
print("Hola")
````

* En la siguiente celda la función `print()` desplegará dos expresiones que corresponden cada una a un objeto de tipo `str`. Cada objeto será desplegado separado por un espacio.

* La segunda expresión (`"Mundo " * 3`) repite el texto tres veces.

* La salida será `Hola Mundo Mundo Mundo `.

````python
print("Hola", "Mundo " * 3)
````

* En la siguiente celda la función ```print()``` desplegará el resultado de una expresión de concatenación entre dos objetos de tipo ```str```. El resultado es un objeto ```str```.

* La salida será ```HolaMundo```.

````python
print("Hola" + "Mundo")
````

* En la siguiente celda la función ```print()``` desplegará tres expresiones que corresponden a:
  * El objeto ```'Tienes'``` de tipo ```str```.
  * El objeto de nombre ```a```.
  * El objeto ```'buenos amigos'``` de tipo ```str```.

Cada expresión será desplegada separada por un espacio.

* La salida será ```Tienes 2 buenos amigos.```.

````python
print("Tienes", a, "buenos amigos.")
````

* En la siguiente celda la función ```print()``` intentará desplegar el resultado de la expresión ```"Tienes" + a + "buenos amigos."```, la cual no es correcta y generará un error de tipo ```TypeError```.

````python
print("Tienes" + a + "buenos amigos.")
````

````python
print("Tienes", end=" ")
print(a, end=" ")
print("buenos amigos.")
````

## Representación de objetos en la salida estándar.

El comportamiento de la salida estándar es distinta entre el intérprete de *Python* y la función `print()` debido al propósito de cada uno.

El propósito del *shell* es informar sobre el resultado de la ejecución de una expresión mediante la representación del objeto, incluyendo además de los resultados, mensajes de error o advertencias, mientras que la función `print()` se utiliza para mostrar información al usuario de manera más amigable y en muchos casos las representaciones del mismo objeto son distintas.

**Nota:** Para fines prácticos, se dirá:
* *Desplegar* para referirse a la función `print()`.
* *Regresar* para referirse a la salida del intérprete de *Python*.
___

**Nota de arquitectura:**

Aunque está fuera del alcance de este curso, cabe hacer notar que los objetos de *Python* implementan métodos especiales para ser representados en la salida estándar.

* El método `__repr__()` es usado por el intérprete.
* El método `__str__()` es usado por la función `print()`.

En muchos casos, las implementaciones para cada método son distintas.

**Ejemplos:**

* En este entorno de *notebook*, cuando una celda termina con una expresión cuyo resultado regresa `None`, el intérprete no despliega salida; en cambio, `print(None)` sí despliega el texto `None`.

````python
None
````

* La función `print()` muestra el texto `None` como representación del objeto `None`.

````python
print(None)
````

* El intérprete muestra la representación de los objetos `str` como texto delimitado por apóstrofes o comillas.

````python
"Hola, mundo."
````

* En el caso de `print()`, se muestra la representación de texto sin los caracteres delimitadores.

````python
print("Hola, mundo.")
````

* El intérprete sólo mostrará el resultado que regrese de la última expresión ejecutada en una celda.

````python
12 + 5
None
max([1, 2, 3])
````

* La función `print()` siempre enviará el resultado a la salida estándar.

````python
print(12 + 5)
print(None)
print(max([1, 2, 3]))
````

* La siguiente celda asignará el resultado de la función `print()` a la variable `valor`. Ese valor es `None`.

````python
valor = print("Este mensaje es enviado a la salida estándar.")
````

* La siguiente celda desplegará el contenido de la variable `valor`, que es `None`.

````python
print(valor)
````

## Despliegue con formato de los objetos ```str```.

Los objetos ```str``` permiten insertar las representaciones de otros objetos dentro de ellos mismos de tres formas:

* El uso del carácter de escape ```%```.
* El método ```format()```, el cual se estudiará más adelante.
* A partir de *Python 3.6* se incluye el uso de *f-strings*, las cuales son descritas en el [*PEP 498*](https://peps.python.org/pep-0498/) y se explorarán más adelante.

**Nota:** El tipo ```str```, las *f-strings* y los *docstrings* se estudiarán con mayor detalle más adelante.

### Inserción de objetos con formato en objetos  `str` mediante el carácter de escape `%`.


Para intercalar valores dentro de un formato específico de texto se utiliza el carácter de sustitución `%`, seguido de algún carácter que definirá el modo de desplegar la expresión correspondiente.

```
"...%<carácter>..." % <expresión 1> 
```

```
"...%<carácter 1>...%<carácter n>..." %(<expresión 1>,...<expresión n>)
```
|Caracter de escape|Modo de despliegue|
|:----------------:|:----------------:|
|`%s`|	cadena de texto|
|`%d`|	 entero|
|`%o`|	 octal|
|`%x`|	 hexadecimal|
|`%f`|	 punto flotante|
|`%e`|	 punto flotante en formato exponencial|    

El uso de `%s`, equivale a aplicar la función `str()` al valor a desplegar.

Para desplegar el signo de porcentaje `%` se utiliza `%%`.

**Ejemplos:**

* La siguiente celda definirán los objetos ```pi``` y ```radio```.

````python
pi = 3.141592
radio = 2
````

* La siguiente celda desplegará un mensaje que incluye:
    * El valor de la variable `radio` como un entero.
    * El resultado de la expresión `2 * radio * pi` como un número de punto flotante.

````python
print("El perímetro de una circunferencia de radio %d es %f." % (radio, 2 * radio * pi))
````

* La siguiente celda desplegará un mensaje que incluye:
    * El valor de la variable `radio` como un entero.
    * El resultado de la expresión `2 * radio * pi` como un entero.

````python
print("El perímetro de una circunferencia de radio %d es %d." % (radio, 2 * radio * pi))
````

* La siguiente celda desplegará un mensaje que incluye:
    * El valor de la variable `radio` como su representación en cadena de caracteres.
    * El resultado de la expresión `2 * radio * pi` como su representación en cadena de caracteres.

````python
print("El perímetro de una circunferencia de radio %s es %s." % (radio, 2 * radio * pi))
````

* La siguiente celda desplegará un mensaje que incluye el valor de la variable `pi` como un número de punto flotante.

````python
print("El valor de pi es %f." % (pi))
````

* La siguiente celda desplegará un mensaje que incluye el valor de la variable `pi` como un número de punto flotante en formato de notación científica.

````python
print("El valor de pi es %e." % pi)
````

La siguiente celda incluye varias expresiones que definen a los objetos ```valor```, ```porciento``` y ```porcentaje```. Dichos objetos serán desplegados dentro de un objeto ```str``` usando el carácter de escape ```%``` y el signo de porcentaje usando ```%%```.

````python
# Esta celda desplegará varias expresiones usando diversos formatos.
valor = 13
porciento = 15
porcentaje = (valor * porciento) / 100
print("El %d%% de %f es %f." % (porciento, valor, porcentaje))
````

#### Despliegue con número fijo de decimales.

Para desplegar un número específico de decimales de un valor de punto flotante, se añade un punto `.` y el número de decimales a desplegarse después del signo de porcentaje `%` y antes del carácter `f` o `e`.

```
"<texto> %.<n>f <texto>"
```

Donde:

```<n>``` es el número de decimales que serán desplegados.

**Ejemplos:**

* La siguiente celda definirán los objetos ```pi``` y ```radio```.

````python
pi = 3.14169265
radio = 2
````

* Las siguientes celdas desplegarán de formas distintas las expresiones ```radio``` y ```2 * pi * radio```.

````python
print("El perímetro de una circunferencia de radio igual a %d es %f." % (radio, 2 * pi * radio))
````

````python
print("El perímetro de una circunferencia de radio igual a %d es %.2f." % (radio, 2 * pi * radio))
````

### Caracteres de escape.

Existen algunos caracteres que por su función o por la sintaxis de *Python*, tales como los apóstrofes, las comillas y los retornos de línea; que deben de utilizar un "carácter de escape", para que puedan ser desplegados. Los caracteres de escape pueden ser introducidos después de una diagonal invertida ```\\```. 

Estos caracteres de escape son usados en conjunto con la función ```print()```.

|Secuencia|Despliegue|
|:-------:|:--------:|
|```\n```    |Retorno de línea|
|```\t```    |Tabulador |
|```\"```    |Comillas  |
|```\'```    |Apóstrofe |
|```\\```   |Diagonal invertida|
|```\xNN```  |Caracter que corresponde al número hexadecimal *NN* en *ASCII*|
|```\uNNNN```  |Caracter que corresponde al número hexadecimal *NN* en *Unicode*|

**Ejemplos:**

* La siguiente celda desplegará un mensaje de texto que incluye caracteres de escape de tabulador y retorno de línea.

````python
print("Primera línea.\nSegunda línea\t con tabulador.")
````

* La siguiente celda desplegará un mensaje de texto que incluye:
    * Los apóstrofes mediante el carácter de escape `\"`.
    * El signo `#` mediante su código hexadecimal en *ASCII* `\x23`.

````python
print("Este es el signo de \"gato\" \x23.")
````

* La siguiente celda desplegará un mensaje de texto que incluye el caracter `ß` usando su número hexadecimal en *Unicode* (`\u00DF`).

````python
print("Beta: \u00DF")
````

* La siguiente celda desplegará un mensaje de texto que incluye el símbolo de corazón `❤` usando su número hexadecimal en *Unicode* (`\u2764`).

````python
print('I \u2764 YOU!')
````

## Entrada estándar con la función ```input()```. 

La función por defecto de entrada estándar para *Python* 3 es ```input()```.

La función ```input()``` captura los caracteres provenientes de entrada estándar (el teclado) hasta que se introduce un retorno de carro <kbd>Intro</kbd> y el contenido capturado regresa al intérprete como una cadena de texto. 

La cadena de caracteres resultante puede ser almacenada como un objeto de tipo ```str``` mediante la asignación de un nombre.

La función permite desplegar un mensaje de tipo ```str``` como parámetro.

``` 
input(<objeto tipo str>)
```

**Ejemplos:**

* La siguiente celda ejecutará a la función ```input()```, la cual pedirá que se ingresen datos y se oprima <kbd>Intro</kbd>. Dicha función regresará la cadena de caracteres capturada como un objeto de tipo ```str```.

````python
input()
````

* La siguiente celda ejecutará a la función ```input()```, la cual pedirá que se ingresen datos y se oprima <kbd>Intro</kbd>. Al objeto ```str``` resultante se le asignará el nombre de ```texto```.

````python
texto = input()
````

* El contenido de la variable `texto` es de tipo `str` (cadena de texto).

````python
type(texto)
````

````python
texto
````

* La siguiente celda ejecutará a la función ```input()```, desplegando el mensaje ```Escribe un nombre: ```, la cual pedirá que se ingresen datos y se oprima <kbd>Intro</kbd>. Al objeto ```str``` resultante se le asignará el nombre de ```nombre```.

````python
nombre = input("Escribe un nombre: ")
````

````python
print(nombre)
````
