# Coincidencia de patrones.

*Python* incorpora la estructura ```match/case``` a partir de la versión 3.10 para realizar coincidencia de patrones de forma clara y legible.

```
match <expresion>:
    case <patron 1>:
        ...
    case <patron 2>:
        ...
    case _ :
        ...
```

Donde:

* ```<expresion>``` es el valor que será evaluado.
* ```<patron i>``` es un patrón contra el que se intentará hacer coincidir a ```<expresion>```.
* ```_``` funciona como comodín cuando no se cumple ningún caso anterior.

**Ejemplo:**

* La siguiente celda evaluará el valor ```3``` mediante la estructura ```match/case```.
* El resultado será ```Tres```.

````python
valor = 3

match valor:
    case 1:
        print("Uno")
    case 2:
        print("Dos")
    case 3:
        print("Tres")
    case _:
        print("Otro numero")
````

## Coincidencia con secuencias.

*Python* permite hacer coincidencia de patrones con secuencias como ```list``` y ```tuple```.

**Ejemplo:**

* La siguiente celda evaluará a la tupla ```("Juan", 25)```.
* El patrón ```(nombre, edad)``` asignará cada elemento a un nombre distinto.

````python
dato = ("Juan", "Pérez", 25)

match dato:
    case (nombre, edad):
        print(f"Nombre: {nombre}, Edad: {edad}")
    case (nombre, apellido, edad):
        print(f"Nombre completo: {nombre + ' ' + apellido}, Edad: {edad}")
````

## Coincidencia con patrones anidados.

Los patrones también pueden descomponer estructuras que contienen otras estructuras en su interior.

**Ejemplo:**

* La siguiente celda evaluará a la tupla ```("Ana", ("CDMX", 30))```.
* El patrón separará el nombre, la ciudad y la edad en una sola operación.

````python
persona = ("Ana", ("CDMX", 30))

match persona:
    case (nombre, (ciudad, edad)):
        print(f"{nombre} vive en {ciudad} y tiene {edad} anos")
````

## Uso de comodines y guardas.

Es posible usar ```_``` como comodín y agregar condiciones adicionales mediante guardas con ```if```.

**Ejemplo:**

* La siguiente celda evaluará a la tupla ```(2, 5)```.
* Primero se intentará verificar si ambos valores son iguales.
* En caso contrario, se ejecutará el patrón general ```(x, y)```.

````python
punto = (2, 2)

match punto:
    case (x, y) if x == y:
        print("Punto en la diagonal")
    case (x, y):
        print(f"Punto: x={x}, y={y}")
````

````python
def datos_personales(*args):
    match args:
        case (nombre, edad):
            print(f"Nombre: {nombre}, Edad: {edad}")
        case (nombre, apellido, edad) if type(nombre) is str and type(apellido) is str:
            print(f"Nombre completo: {nombre + ' ' + apellido}, Edad: {edad}")
        case _:
            print("Estructura inválida.")        
````

````python
datos_personales("Juan", 'Pérez', 22)
````

````python
datos_personales("Juan", True, 22)
````

````python
datos_personales(1, 2)
````

## Coincidencia con valores literales.

También es posible usar ```match/case``` para comparar valores literales como cadenas de caracteres.

**Ejemplo:**

* La siguiente celda evaluará a la cadena ```"rojo"```.
* El resultado será ```El fuego es rojo```.

**Notas:**

* Conviene usar ```match/case``` cuando los patrones describen estructuras o alternativas claras.
* Esta estructura no sustituye a las expresiones regulares cuando se trabaja con texto libre.
* ```match/case``` sólo está disponible desde *Python* 3.10.

````python
color = "azulado"

match color.lower():
    case "azul" | "azulado":
        print("El cielo es azul")
    case "rojo":
        print("El fuego es rojo")
    case _:
        print("Color desconocido")
````

## Coincidencia con diccionarios.

La estructura ```match/case``` también permite hacer coincidencia con claves y valores de objetos ```dict```.

**Ejemplo:**

* La siguiente celda evaluará al diccionario ```{"nombre": "Ana", "edad": 30}```.
* Si las claves esperadas están presentes, el patrón extraerá sus valores.

````python
def datos_personales_dict(**kwargs):
    match kwargs:
        case {"nombre": nombre, "edad": edad}:
            print(f"Nombre: {nombre}, Edad: {edad}")
        case _:
            print("Patron no reconocido")
````

````python
datos_personales_dict(nombre="Ana", edad=30)
````

````python
datos_personales_dict(2, 3)
````

````python
datos_personales_dict(nombre='Juan', apellido='Martínez', edad=27)
````

## Resumen.

* ```match/case``` permite realizar coincidencia de patrones sobre valores y estructuras.
* Es especialmente util cuando se trabaja con secuencias, diccionarios y patrones anidados.
* Las guardas con ```if``` permiten agregar condiciones adicionales a cada caso.
