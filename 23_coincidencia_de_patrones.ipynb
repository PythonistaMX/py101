# Coincidencia de Patrones en Python (match/case)

En este cuaderno aprenderás a utilizar la estructura `match/case` introducida en Python 3.10 para realizar coincidencia de patrones de manera clara y legible.

> **Nota:** Las expresiones regulares (`re`) se estudian a profundidad en el curso py121. Aquí nos enfocamos en el uso de `match/case` para patrones estructurados.

```python
# Ejemplo básico de match/case

valor = 3

match valor:
    case 1:
        print("Uno")
    case 2:
        print("Dos")
    case 3:
        print("Tres")
    case _:
        print("Otro número")
```

## Coincidencia de patrones con estructuras

`match/case` permite trabajar con estructuras como listas, tuplas o diccionarios de forma muy legible.

```python
# Coincidencia con listas y tuplas

dato = ("Juan", 25)

match dato:
    case (nombre, edad):
        print(f"Nombre: {nombre}, Edad: {edad}")
```

## Coincidencia con patrones anidados

Puedes usar `match/case` para descomponer estructuras anidadas fácilmente.

```python
# Coincidencia con estructuras anidadas
persona = ("Ana", ("CDMX", 30))

match persona:
    case (nombre, (ciudad, edad)):
        print(f"{nombre} vive en {ciudad} y tiene {edad} años")
```

## Uso de comodines y guardas

Puedes usar `_` como comodín y agregar condiciones extra (guardas) con `if`.

```python
# Uso de comodines y guardas
punto = (2, 5)

match punto:
    case (x, y) if x == y:
        print("Punto en la diagonal")
    case (x, y):
        print(f"Punto: x={x}, y={y}")
```

## Buenas prácticas y limitaciones

- Usa `match/case` para estructuras de datos claras y legibles.
- No reemplaza a las expresiones regulares para coincidencia en texto libre.
- Recuerda: sólo disponible desde Python 3.10+.

```python
# Ejercicio: ¿Qué imprime este código?
color = "rojo"

match color:
    case "azul":
        print("El cielo es azul")
    case "rojo":
        print("El fuego es rojo")
    case _:
        print("Color desconocido")
```

## Resumen

- `match/case` permite coincidencia estructurada y legible.
- Para coincidencia de patrones en texto, usaremos expresiones regulares en py121.
